## What is an Operator?

An operator is a symbol that tells Python to perform a specific operation on one or more values (called operands).

```
10 + 5 = 15
^^   ^   ^^
operand  operator  result
```

## 1. Arithmetic Operators

**Definition:** Perform basic mathematical calculations.

| Operator | Symbol | Example |
|----------|--------|---------|
| Addition | + | 10 + 3 = 13 |
| Subtraction | - | 10 - 3 = 7 |
| Multiplication | * | 10 * 3 = 30 |
| Division | / | 10 / 3 = 3.33 |
| Floor Division | // | 10 // 3 = 3 |
| Modulus | % | 10 % 3 = 1 |
| Exponent | ** | 10 ** 3 = 1000 |

**Use Cases:**
- Salary calculations
- Tax computations
- Discount pricing
- Scientific formulas

````python
a = 10
b = 3

print("Addition       :", a + b)
print("Subtraction    :", a - b)
print("Multiplication :", a * b)
print("Division       :", a / b)
print("Floor Division :", a // b)
print("Modulus        :", a % b)
print("Exponent       :", a ** b)
````

````python
# Real World: Salary Calculation
basic_salary = 50000
bonus = 5000
tax_rate = 0.10

gross = basic_salary + bonus
tax = gross * tax_rate
net_salary = gross - tax

print("Gross Salary :", gross)
print("Tax Deducted :", tax)
print("Net Salary   :", net_salary)
````

## 2. Comparison Operators

**Definition:** Compare two values and return `True` or `False`.

| Operator | Symbol | Meaning |
|----------|--------|---------|
| Equal | == | Are both values equal? |
| Not Equal | != | Are the values different? |
| Greater Than | > | Is left greater than right? |
| Less Than | < | Is left less than right? |
| Greater or Equal | >= | Is left >= right? |
| Less or Equal | <= | Is left <= right? |

**Use Cases:**
- Age validation
- Eligibility checks
- Filtering records
- Pass/fail decisions

````python
a = 10
b = 20

print("Equal            :", a == b)
print("Not Equal        :", a != b)
print("Greater Than     :", a > b)
print("Less Than        :", a < b)
print("Greater or Equal :", a >= b)
print("Less or Equal    :", a <= b)
````

````python
# Real World: Student Pass/Fail
marks = 72
pass_mark = 50

if marks >= pass_mark:
    print("Result: Pass")
else:
    print("Result: Fail")
````

## 3. Logical Operators

**Definition:** Combine multiple conditions and return `True` or `False`.

| Operator | Meaning | Example |
|----------|---------|--------|
| and | Both conditions must be True | age > 18 and has_id == True |
| or | At least one condition must be True | is_admin or is_manager |
| not | Reverses the condition | not is_blocked |

**Truth Table:**

| A | B | A and B | A or B |
|---|---|---------|--------|
| True | True | True | True |
| True | False | False | True |
| False | True | False | True |
| False | False | False | False |

**Use Cases:**
- Login validation
- Access control
- Data filtering
- Business rules

````python
x = True
y = False

print("x and y :", x and y)
print("x or y  :", x or y)
print("not x   :", not x)
````

````python
# Real World: Loan Eligibility
age = 30
salary = 60000
credit_score = 750

eligible = age >= 21 and salary >= 50000 and credit_score >= 700

if eligible:
    print("Loan Approved")
else:
    print("Loan Rejected")
````

## 4. Assignment Operators

**Definition:** Assign values to variables, often combining with an arithmetic operation.

| Operator | Example | Equivalent To |
|----------|---------|---------------|
| = | x = 10 | Assign 10 |
| += | x += 5 | x = x + 5 |
| -= | x -= 5 | x = x - 5 |
| *= | x *= 2 | x = x * 2 |
| /= | x /= 2 | x = x / 2 |
| //= | x //= 2 | x = x // 2 |
| %= | x %= 3 | x = x % 3 |
| **= | x **= 2 | x = x ** 2 |

**Use Cases:**
- Running totals
- Counters
- Accumulating values in loops

````python
x = 10
print("Initial :", x)

x += 5
print("After += 5 :", x)

x -= 3
print("After -= 3 :", x)

x *= 2
print("After *= 2 :", x)

x //= 4
print("After //= 4:", x)
````

````python
# Real World: Running Total of Sales
sales = [1500, 2300, 1800, 3200, 2700]
total = 0

for amount in sales:
    total += amount

print("Total Sales :", total)
````

## 5. Identity Operators

**Definition:** Check whether two variables point to the **same object in memory** (not just equal values).

| Operator | Meaning |
|----------|---------|
| is | Returns True if both variables are the same object |
| is not | Returns True if both variables are different objects |

> `==` checks **value equality**  
> `is` checks **memory identity** (same object)

**Use Cases:**
- Checking for `None`
- Singleton comparisons
- Object reference checks

````python
a = [1, 2, 3]
b = a           # b points to the same object as a
c = [1, 2, 3]   # c is a new object with same values

print("a is b  :", a is b)    # True  - same object
print("a is c  :", a is c)    # False - different objects
print("a == c  :", a == c)    # True  - same values
````

````python
# Real World: Checking for missing data
customer_email = None

if customer_email is None:
    print("Email is missing - cannot send notification")
else:
    print("Sending email to:", customer_email)
````

## 6. Membership Operators

**Definition:** Check whether a value exists inside a sequence (list, tuple, string, set, dict).

| Operator | Meaning |
|----------|---------|
| in | Returns True if value is found in the sequence |
| not in | Returns True if value is NOT found in the sequence |

**Use Cases:**
- Searching records
- Allowlist / blocklist checks
- Validating user input
- Filtering data

````python
fruits = ["apple", "banana", "cherry"]

print("banana in fruits    :", "banana" in fruits)
print("mango in fruits     :", "mango" in fruits)
print("mango not in fruits :", "mango" not in fruits)

# Works on strings too
message = "Hello, World!"
print("'World' in message  :", "World" in message)
````

````python
# Real World: Blocked Users Check
blocked_users = ["spam_user", "bot123", "hacker99"]
current_user = "hacker99"

if current_user in blocked_users:
    print("Access Denied:", current_user)
else:
    print("Welcome:", current_user)
````

## 7. Bitwise Operators

**Definition:** Operate on binary (bit-level) representations of integers.

| Operator | Symbol | Meaning |
|----------|--------|---------|
| AND | & | 1 only if both bits are 1 |
| OR | \| | 1 if at least one bit is 1 |
| XOR | ^ | 1 if bits are different |
| NOT | ~ | Flips all bits |
| Left Shift | << | Shifts bits left (multiplies by 2) |
| Right Shift | >> | Shifts bits right (divides by 2) |

**Use Cases:**
- Networking (subnet masks)
- Encryption
- Permission flags
- Performance optimisation

````python
a = 10   # binary: 1010
b = 4    # binary: 0100

print("a & b  (AND)        :", a & b)   # 0000 = 0
print("a | b  (OR)         :", a | b)   # 1110 = 14
print("a ^ b  (XOR)        :", a ^ b)   # 1110 = 14
print("~a     (NOT)        :", ~a)      # -11
print("a << 2 (Left Shift) :", a << 2)  # 40  (10 * 4)
print("a >> 2 (Right Shift):", a >> 2)  # 2   (10 / 4)
````

````python
# Real World: Permission Flags
READ    = 0b001   # 1
WRITE   = 0b010   # 2
EXECUTE = 0b100   # 4

user_permissions = READ | WRITE   # Grant read + write

print("Can Read    :", bool(user_permissions & READ))
print("Can Write   :", bool(user_permissions & WRITE))
print("Can Execute :", bool(user_permissions & EXECUTE))
````

## 8. Operator Precedence

**Definition:** The order in which Python evaluates operators in an expression. Higher precedence operators are evaluated first.

| Priority | Operator | Description |
|----------|----------|-------------|
| 1 (Highest) | () | Parentheses |
| 2 | ** | Exponentiation |
| 3 | ~, +, - | Unary operators |
| 4 | *, /, //, % | Multiplication, Division |
| 5 | +, - | Addition, Subtraction |
| 6 | <<, >> | Bitwise Shift |
| 7 | & | Bitwise AND |
| 8 | ^ | Bitwise XOR |
| 9 | \| | Bitwise OR |
| 10 | ==, !=, <, >, <=, >= | Comparison |
| 11 | not | Logical NOT |
| 12 | and | Logical AND |
| 13 (Lowest) | or | Logical OR |

> **Tip:** When in doubt, use parentheses to make your intention explicit.

````python
# Without parentheses - follows precedence rules
result1 = 10 + 5 * 2
print("10 + 5 * 2  =", result1)   # 20 (not 30)

# With parentheses - overrides precedence
result2 = (10 + 5) * 2
print("(10 + 5) * 2 =", result2)  # 30

# Exponent before multiplication
result3 = 2 ** 3 * 4
print("2 ** 3 * 4  =", result3)   # 32 (8 * 4)

# Comparison before logical
result4 = 5 > 3 and 10 < 20
print("5>3 and 10<20 =", result4) # True
````
