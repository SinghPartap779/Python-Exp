# Python Set Operations Cheat Sheet

## 1. Creating Sets
Sets are **unordered** collections of **unique** elements. Use `{}` or `set()`.

```python
s1 = {1, 2, 3, 4}
s2 = set([3, 4, 5, 6])   # from a list
s3 = set("hello")        # from a string -> unique chars
empty = set()            # NOT {} — that creates a dict!

print(s1)        # {1, 2, 3, 4}
print(s2)        # {3, 4, 5, 6}
print(s3)        # {"h", "e", "l", "o"}  (duplicates removed)
print(len(s1))   # 4
```

## 2. Adding & Removing Elements
`add()`, `update()`, `remove()`, `discard()`, `pop()`, `clear()`

```python
s = {1, 2, 3}

s.add(4)              # add single element
print(s)              # {1, 2, 3, 4}

s.update([5, 6])      # add multiple elements
print(s)              # {1, 2, 3, 4, 5, 6}

s.remove(6)           # raises KeyError if not found
s.discard(99)         # safe — no error if missing
print(s)              # {1, 2, 3, 4, 5}

popped = s.pop()      # removes & returns an arbitrary element
print(popped)

s.clear()             # empties the set
print(s)              # set()
```

## 3. Set Operations (Math)
Union, Intersection, Difference, Symmetric Difference

```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

# Union — all elements from both
print(A | B)            # {1, 2, 3, 4, 5, 6}
print(A.union(B))       # same

# Intersection — only common elements
print(A & B)                   # {3, 4}
print(A.intersection(B))       # same

# Difference — in A but not in B
print(A - B)                   # {1, 2}
print(A.difference(B))         # same

# Symmetric Difference — in either but not both
print(A ^ B)                            # {1, 2, 5, 6}
print(A.symmetric_difference(B))        # same
```

## 4. In-place Set Operations
Modify the set directly without creating a new one.

```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

A |= B   # A = A.union(B)
print(A) # {1, 2, 3, 4, 5, 6}

A = {1, 2, 3, 4}
A &= B   # A = A.intersection(B)
print(A) # {3, 4}

A = {1, 2, 3, 4}
A -= B   # A = A.difference(B)
print(A) # {1, 2}

A = {1, 2, 3, 4}
A ^= B   # A = A.symmetric_difference(B)
print(A) # {1, 2, 5, 6}
```

## 5. Membership & Relationship Checks
`in`, `issubset()`, `issuperset()`, `isdisjoint()`

```python
A = {1, 2, 3, 4, 5}
B = {2, 3}
C = {6, 7}

print(3 in A)             # True  — membership test
print(9 not in A)         # True

print(B.issubset(A))      # True  — B is contained in A
print(B <= A)             # same

print(A.issuperset(B))    # True  — A contains all of B
print(A >= B)             # same

print(A.isdisjoint(C))    # True  — no common elements
```

## 6. Copying Sets
`copy()` creates a shallow copy so changes don't affect the original.

```python
original = {1, 2, 3}

alias = original         # same object — changes affect original
clone = original.copy()  # independent copy

alias.add(99)
clone.add(100)

print(original)  # {1, 2, 3, 99}  — alias changed it
print(clone)     # {1, 2, 3, 100} — unaffected
```

## 7. Frozenset
An **immutable** set — can be used as a dictionary key or inside another set.

```python
fs = frozenset([1, 2, 3, 2, 1])
print(fs)              # frozenset({1, 2, 3})

# Supports all read operations
print(2 in fs)         # True
print(fs & {2, 3, 4})  # frozenset({2, 3})

# Cannot be modified
# fs.add(4)  -> AttributeError

# Useful as a dict key
d = {frozenset({1, 2}): "pair"}
print(d[frozenset({1, 2})])  # pair
```

## 8. Set Comprehension
Build sets concisely using comprehension syntax.

```python
squares = {x**2 for x in range(1, 6)}
print(squares)   # {1, 4, 9, 16, 25}

evens = {x for x in range(10) if x % 2 == 0}
print(evens)     # {0, 2, 4, 6, 8}

# Deduplicate a list while keeping unique values
nums = [1, 2, 2, 3, 3, 3, 4]
unique = {x for x in nums}
print(unique)    # {1, 2, 3, 4}
```

## Practice Challenges

### Challenge 1: Remove Duplicates
Remove all duplicates from `[1, 2, 2, 3, 4, 4, 5]` using a set.

```python
nums = [1, 2, 2, 3, 4, 4, 5]
unique = list(set(nums))
print(unique)  # [1, 2, 3, 4, 5]  (order may vary)
```

### Challenge 2: Common Elements
Find common elements between `{10, 20, 30, 40}` and `{30, 40, 50, 60}`.

```python
A = {10, 20, 30, 40}
B = {30, 40, 50, 60}
common = A & B
print(common)  # {30, 40}
```

### Challenge 3: Unique Words
Count how many unique words are in `"the cat sat on the mat the cat"`.

```python
sentence = "the cat sat on the mat the cat"
unique_words = set(sentence.split())
print(unique_words)       # {"the", "cat", "sat", "on", "mat"}
print(len(unique_words))  # 5
```

### Challenge 4: Subset Check
Check if `{2, 4}` is a subset of `{1, 2, 3, 4, 5}`.

```python
small = {2, 4}
big   = {1, 2, 3, 4, 5}
print(small.issubset(big))  # True
print(small <= big)         # True
```

### Challenge 5: Symmetric Difference
Find elements that are in `{1, 2, 3, 4}` or `{3, 4, 5, 6}` but **not in both**.

```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}
result = A ^ B
print(result)  # {1, 2, 5, 6}
```
