Numbers - Integer, Float - Whole numbers without a decimal point, positive or negative. Example: 10, -5, 2026

```python
print("=== Numbers ===")
age = 25                # int
pi = 3.14159            # float
z = 2 + 3j              # complex
print("Integer:", age, type(age))
print("Float:", pi, type(pi))
print("Complex:", z, type(z))
```

Strings - A sequence of characters enclosed in single, double, or triple quotes. Example: "Hello", 'Python', """Multiline text"""

```python
print("\n=== Strings ===")
name = "Alice"
greeting = 'Hello'
print("String:", name, type(name))
print("Uppercase:", name.upper())
print("Repeat:", name * 2)
```

Complex Numbers - Numbers with a real and an imaginary part, written as a + bj.
[Use Case] - Electrical Engineering (AC Circuits) - Complex numbers represent impedance (resistance + reactance).

```python
# Complex number example
z1 = 2 + 3j   # complex number with real part 2 and imaginary part 3
z2 = 1 + 4j   # another complex number

# Operations
print("z1:", z1)
print("z2:", z2)
print("Addition:", z1 + z2)       # (2+3j) + (1+4j) = (3+7j)
print("Multiplication:", z1 * z2) # (2+3j) * (1+4j) = (-10+11j)

# Accessing parts
print("Real part of z1:", z1.real)
print("Imaginary part of z1:", z1.imag)
```

```python
# Impedance: R + jX
#Complex numbers represent impedance (resistance + reactance).
resistance = 50
reactance = 30
impedance = complex(resistance, reactance)
print("Impedance:", impedance)
```

```python
import cmath
#Complex numbers are used in Fourier Transforms to analyze signals.
# Represent a wave as a complex exponential
angle = cmath.pi / 4
wave = cmath.exp(1j * angle)
print("Wave:", wave)
```

Booleans - Logical values representing truth — either True or False.

```python
print("\n=== Booleans ===")
is_student = True
passed_exam = False
print("Boolean True:", is_student, type(is_student))
print("Boolean False:", passed_exam, type(passed_exam))
print("Comparison (5 > 3):", 5 > 3)
```

**Lists**  
- Ordered, mutable collection of items (can mix data types).
- Mutable = Changeable after creation.
- Immutable = Cannot be changed after creation.

**Mutable Collections:** `list`, `dict`, `set`  
**Immutable Collections:** `tuple`, `frozenset`, `str`

```python
print("\n=== Lists ===")
fruits = ["apple", "banana", "cherry"]
fruits.append("mango")
print("List:", fruits, type(fruits))
```

```python
#Example of List (Mutable)
# A list is mutable
fruits = ["apple", "banana", "cherry"]
print("Original:", fruits)

# Modify the list
fruits.append("mango")       # Add new item
fruits[1] = "blueberry"      # Update existing item
fruits.remove("apple")       # Remove item

print("Modified:", fruits)
```

Tuples - Definition: An ordered, immutable collection of items. Example: (10, 20, 30)

```python
print("\n=== Tuples ===")
coordinates = (10, 20)
print("Tuple:", coordinates, type(coordinates))
```

Range - A sequence of numbers, often used in loops. Example: range(5) → 0,1,2,3,4

```python
print("\n=== Range ===")
numbers = range(5)
print("Range:", list(numbers), type(numbers))
```

Sets - An unordered collection of unique items. Example: {"red", "blue", "green"}

```python
print("\n=== Sets ===")
colors = {"red", "blue", "green"}
colors.add("yellow")
print("Set:", colors, type(colors))
```

Frozen Set - An immutable version of a set (cannot be changed after creation). Example: frozenset([1,2,3])

```python
print("\n=== Frozen Set ===")
frozen_colors = frozenset(["red", "blue", "green"])
```

Dictionary (Dict) - A collection of key-value pairs, where keys are unique.

```python
# Creating a dictionary
student = {
    "name": "Alice",
    "age": 25,
    "is_student": True
}

# Accessing values by key
print("Name:", student["name"])       # Alice
print("Age:", student["age"])         # 25

# Adding a new key-value pair
student["course"] = "Python Basics"

# Updating a value
student["age"] = 26

# Removing a key-value pair
del student["is_student"]

# Iterating through dictionary
for key, value in student.items():
    print(key, ":", value)
```
