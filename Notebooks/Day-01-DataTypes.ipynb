## Numbers — int · float · complex
**int:** whole numbers (no decimal) · **float:** decimal numbers · **complex:** real + imaginary part (a+bj)

```python
print("=== Numbers ===")
age = 25                # int
pi = 3.14159            # float
z = 2 + 3j              # complex
print("Integer:", age, type(age))
print("Float:", pi, type(pi))
print("Complex:", z, type(z))

# Type conversions
print("\n--- Type Conversions ---")
print("int('42')  :", int('42'))      # str → int
print("float(10)  :", float(10))      # int → float
print("str(3.14)  :", str(3.14))      # float → str
print("bool(0)    :", bool(0))        # 0 is False
print("bool('hi') :", bool('hi'))     # non-empty str is True
```

## Strings
Ordered sequence of characters. Enclosed in `'single'`, `"double"`, or `"""triple"""` quotes. Immutable.

```python
print("=== Strings ===")
name = "Alice"
message = "Hello, World!"

# Common string methods
print("Upper      :", name.upper())
print("Lower      :", name.lower())
print("Length     :", len(message))
print("Replace    :", message.replace("World", "Python"))
print("Split      :", message.split(", "))
print("Strip      :", "  hello  ".strip())
print("Starts with:", message.startswith("Hello"))
print("f-string   :", f"My name is {name} and I am learning Python")
```

## Complex Numbers
Form: `a + bj` where `a` = real part, `b` = imaginary part.
**Use:** AC circuits · signal processing · Fourier transforms

```python
# Complex number example
z1 = 2 + 3j   # complex number with real part 2 and imaginary part 3
z2 = 1 + 4j   # another complex number

# Operations
print("z1:", z1)
print("z2:", z2)
print("Addition:", z1 + z2)       # (2+3j) + (1+4j) = (3+7j)
print("Multiplication:", z1 * z2) # (2+3j) * (1+4j) = (-10+11j)

# Accessing parts
print("Real part of z1:", z1.real)
print("Imaginary part of z1:", z1.imag)
```

```python
# Impedance: R + jX
#Complex numbers represent impedance (resistance + reactance).
resistance = 50
reactance = 30
impedance = complex(resistance, reactance)
print("Impedance:", impedance)
```

```python
import cmath
#Complex numbers are used in Fourier Transforms to analyze signals.
# Represent a wave as a complex exponential
angle = cmath.pi / 4
wave = cmath.exp(1j * angle)
print("Wave:", wave)
```

## Booleans
Only two values: `True` or `False`. Result of comparisons and logical operations.
**Use:** conditions · flags · filtering · validation

```python
print("\n=== Booleans ===")
is_student = True
passed_exam = False
print("Boolean True:", is_student, type(is_student))
print("Boolean False:", passed_exam, type(passed_exam))
print("Comparison (5 > 3):", 5 > 3)
```

## Lists
Ordered, **mutable** collection — items can be added, removed, or changed after creation.
`list` · `dict` · `set` → mutable | `tuple` · `frozenset` · `str` → immutable

```python
print("\n=== Lists ===")
fruits = ["apple", "banana", "cherry"]
fruits.append("mango")
print("List:", fruits, type(fruits))
```

```python
#Example of List (Mutable)
# A list is mutable
fruits = ["apple", "banana", "cherry"]
print("Original:", fruits)

# Modify the list
fruits.append("mango")       # Add new item
fruits[1] = "blueberry"      # Update existing item
fruits.remove("apple")       # Remove item

print("Modified:", fruits)
```

## Tuples
Ordered, **immutable** collection. Use when data should not change.
**Use:** coordinates · RGB colours · database rows · function return values

```python
print("\n=== Tuples ===")
coordinates = (10, 20)
print("Tuple:", coordinates, type(coordinates))
```

## Range
Generates a sequence of integers on demand (memory-efficient). Used mainly in loops.
`range(stop)` · `range(start, stop)` · `range(start, stop, step)`

```python
print("\n=== Range ===")
numbers = range(5)
print("Range:", list(numbers), type(numbers))
```

## Sets
Unordered collection of **unique** items. Automatically removes duplicates.
**Use:** deduplication · membership checks · set operations (union, intersection)

```python
print("\n=== Sets ===")
colors = {"red", "blue", "green"}
colors.add("yellow")
print("Set:", colors, type(colors))
```

## Frozen Set
Immutable version of a set — cannot be modified after creation.
**Use:** dictionary keys · hashable sets · fixed lookup tables

```python
print("\n=== Frozen Set ===")
frozen_colors = frozenset(["red", "blue", "green"])
print("Frozen Set:", frozen_colors, type(frozen_colors))

# Frozen sets support set operations but cannot be modified
A = frozenset([1, 2, 3])
B = frozenset([2, 3, 4])
print("Union:", A | B)
print("Intersection:", A & B)
```

## Dictionary (dict)
Key-value pairs. Keys are unique. Ordered (Python 3.7+). O(1) lookup.
**Use:** configs · lookup tables · JSON data · counting · grouping

```python
# Creating a dictionary
student = {
    "name": "Alice",
    "age": 25,
    "is_student": True
}

# Accessing values by key
print("Name:", student["name"])       # Alice
print("Age:", student["age"])         # 25

# Adding a new key-value pair
student["course"] = "Python Basics"

# Updating a value
student["age"] = 26

# Removing a key-value pair
del student["is_student"]

# Iterating through dictionary
for key, value in student.items():
    print(key, ":", value)
```
