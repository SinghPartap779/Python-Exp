# Python Regular Expressions Cheat Sheet

## 1. What are Regular Expressions?

A **regular expression (regex)** is a pattern string used to search, match, and manipulate text.

Python's `re` module provides all regex tools.

## 2. Key Regex Patterns

| Pattern | Matches |
|---|---|
| `.` | Any character except newline |
| `\d` | Digit 0-9 |
| `\D` | Non-digit |
| `\w` | Word character (a-z, A-Z, 0-9, _) |
| `\W` | Non-word character |
| `\s` | Whitespace (space, tab, newline) |
| `\S` | Non-whitespace |
| `^` | Start of string |
| `$` | End of string |
| `*` | 0 or more |
| `+` | 1 or more |
| `?` | 0 or 1 (optional) |
| `{n}` | Exactly n times |
| `{n,m}` | Between n and m times |
| `[abc]` | One of a, b, c |
| `[^abc]` | Not a, b, or c |
| `(abc)` | Capture group |
| `a\|b` | a or b |

## 3. Core Functions

```python
import re

text = "My phone is 07712-345678 and email is alice@example.com"

# re.search() — find first match anywhere in string
match = re.search(r"\d{5}-\d{6}", text)
if match:
    print("Found:", match.group())   # 07712-345678

# re.match() — match at the START of the string only
print(re.match(r"My", text))         # Match object
print(re.match(r"phone", text))      # None
```

```python
# re.findall() — return list of all matches
sentence = "cat bat hat mat rat"
matches = re.findall(r"[cbhm]at", sentence)
print(matches)   # ['cat', 'bat', 'hat', 'mat']
```

```python
# re.sub() — replace matches
text = "Hello   World   Python"
cleaned = re.sub(r"\s+", " ", text)   # collapse multiple spaces
print(cleaned)   # Hello World Python
```

```python
# re.split() — split on a pattern
data = "one,two;three|four"
parts = re.split(r"[,;|]", data)
print(parts)   # ['one', 'two', 'three', 'four']
```

## 4. Capture Groups

```python
# Extract parts of a date
date_str = "Today is 2024-06-15"
match = re.search(r"(\d{4})-(\d{2})-(\d{2})", date_str)
if match:
    print("Full:",  match.group(0))   # 2024-06-15
    print("Year:",  match.group(1))   # 2024
    print("Month:", match.group(2))   # 06
    print("Day:",   match.group(3))   # 15
```

```python
# Named groups — more readable
match = re.search(r"(?P<year>\d{4})-(?P<month>\d{2})-(?P<day>\d{2})", date_str)
if match:
    print(match.group("year"))   # 2024
    print(match.groupdict())     # {'year': '2024', 'month': '06', 'day': '15'}
```

## 5. Compiling Patterns

If you reuse a pattern many times, compile it first for speed.

```python
email_pattern = re.compile(r"[\w.+-]+@[\w-]+\.[\w.]+")

emails = ["alice@example.com", "not-an-email", "bob.smith@work.org"]
for text in emails:
    if email_pattern.search(text):
        print(f"Valid email: {text}")
    else:
        print(f"Not an email: {text}")
```

## 6. Flags

| Flag | Effect |
|---|---|
| `re.IGNORECASE` | Case-insensitive matching |
| `re.MULTILINE` | `^` and `$` match line boundaries |
| `re.DOTALL` | `.` matches newlines too |

```python
text = "Python is great. PYTHON rocks. python forever."
matches = re.findall(r"python", text, re.IGNORECASE)
print(matches)   # ['Python', 'PYTHON', 'python']
```

### Challenge 1: Email Validator
Check if a string is a valid email address.

```python
def is_valid_email(email):
    pattern = r"^[\w.+-]+@[\w-]+\.[a-zA-Z]{2,}$"
    return bool(re.match(pattern, email))

tests = ["user@example.com", "bad@", "@nodomain.com", "valid.email+tag@sub.domain.org"]
for e in tests:
    print(f"{e:35} -> {'Valid' if is_valid_email(e) else 'Invalid'}")
```

### Challenge 2: Phone Number Extractor
Extract all phone numbers from a block of text.

```python
text = """
Call us at 0800-123-4567 or 07700 900123.
International: +44-20-7946-0958
Also: (020) 7946-9000
"""
phones = re.findall(r"[\+\(]?[\d\s\-\(\)]{7,20}", text)
print([p.strip() for p in phones if len(p.strip()) >= 7])
```

### Challenge 3: Password Strength Checker
Use regex to check if a password meets: 8+ chars, one uppercase, one digit, one special char.

```python
def check_password(pwd):
    rules = [
        (r".{8,}",           "At least 8 characters"),
        (r"[A-Z]",           "One uppercase letter"),
        (r"[0-9]",           "One digit"),
        (r"[!@#$%^&*()_+]", "One special character"),
    ]
    results = []
    for pattern, rule in rules:
        ok = bool(re.search(pattern, pwd))
        results.append((rule, ok))
    return results

for name, pwd in [("Weak", "abc"), ("Strong", "Secure@123")]:
    print(f"\n{name} password: '{pwd}'")
    for rule, passed in check_password(pwd):
        print(f"  {'OK' if passed else 'FAIL'} — {rule}")
```

### Challenge 4: Log File Parser
Extract timestamps and log levels from log lines.

```python
logs = """2024-06-01 10:23:45 ERROR  Database connection failed
2024-06-01 10:23:46 INFO   Retrying connection...
2024-06-01 10:23:47 INFO   Connected successfully
2024-06-01 10:24:00 WARNING Slow query detected"""

pattern = re.compile(r"(\d{4}-\d{2}-\d{2} \d{2}:\d{2}:\d{2}) (\w+)\s+(.+)")
for match in pattern.finditer(logs):
    ts, level, message = match.groups()
    print(f"[{level}] {ts} — {message}")
```
