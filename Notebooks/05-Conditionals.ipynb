# Python Conditionals Cheat Sheet

## 1. What are Conditionals?

Conditionals let your program **make decisions** — run one block of code or another depending on whether a condition is `True` or `False`.

## 2. if Statement

```python
age = 18

if age >= 18:
    print("You are an adult.")
```

## 3. if / else

```python
score = 45

if score >= 50:
    print("Pass")
else:
    print("Fail")
```

## 4. if / elif / else

Use `elif` to check multiple conditions in sequence. Only the **first** matching block runs.

```python
grade = 75

if grade >= 90:
    print("A")
elif grade >= 80:
    print("B")
elif grade >= 70:
    print("C")
elif grade >= 60:
    print("D")
else:
    print("F")
```

## 5. Nested Conditionals

An `if` inside another `if`. Use sparingly — deep nesting is hard to read.

```python
logged_in = True
is_admin = False

if logged_in:
    if is_admin:
        print("Welcome, Admin!")
    else:
        print("Welcome, User!")
else:
    print("Please log in.")
```

## 6. Ternary Operator (One-Line if/else)

Syntax: `value_if_true if condition else value_if_false`

```python
age = 20
status = "adult" if age >= 18 else "minor"
print(status)   # adult

# Also works inline
print("Even" if 10 % 2 == 0 else "Odd")
```

## 7. Truthy and Falsy Values

Python treats these as `False`:
- `0`, `0.0`, `""`, `[]`, `{}`, `()`, `None`, `False`

Everything else is `True`.

```python
values = [0, 1, "", "hello", [], [1,2], None, True]

for v in values:
    print(f"{repr(v):12} -> {'Truthy' if v else 'Falsy'}")
```

## 8. Comparison Operators in Conditions

| Operator | Meaning |
|---|---|
| `==` | Equal |
| `!=` | Not equal |
| `>` | Greater than |
| `<` | Less than |
| `>=` | Greater or equal |
| `<=` | Less or equal |
| `is` | Same object in memory |
| `in` | Member of sequence |

```python
fruits = ["apple", "banana", "cherry"]

if "banana" in fruits:
    print("Found banana!")

if len(fruits) == 3:
    print("Exactly 3 fruits.")
```

## 9. Logical Operators: and, or, not

```python
age = 25
has_id = True

# and: both must be True
if age >= 18 and has_id:
    print("Entry allowed")

# or: at least one must be True
temperature = 5
if temperature < 0 or temperature > 35:
    print("Extreme weather!")

# not: flips the boolean
raining = False
if not raining:
    print("No umbrella needed.")
```

## 10. match Statement (Python 3.10+)

Like a switch/case from other languages.

```python
day = "Monday"

match day:
    case "Saturday" | "Sunday":
        print("Weekend!")
    case "Monday":
        print("Start of the week.")
    case _:
        print("Weekday.")
```

### Challenge 1: FizzBuzz
Print numbers 1–30. For multiples of 3 print `Fizz`, multiples of 5 print `Buzz`, both print `FizzBuzz`.

```python
for n in range(1, 31):
    if n % 15 == 0:
        print("FizzBuzz")
    elif n % 3 == 0:
        print("Fizz")
    elif n % 5 == 0:
        print("Buzz")
    else:
        print(n)
```

### Challenge 2: BMI Calculator
Ask for weight (kg) and height (m), calculate BMI, and print the category.

```python
weight = float(input("Weight in kg: "))
height = float(input("Height in m: "))
bmi = weight / (height ** 2)

if bmi < 18.5:
    category = "Underweight"
elif bmi < 25:
    category = "Normal"
elif bmi < 30:
    category = "Overweight"
else:
    category = "Obese"

print(f"BMI: {bmi:.1f} — {category}")
```

### Challenge 3: Leap Year Checker
A year is a leap year if divisible by 4, except centuries unless divisible by 400.

```python
def is_leap(year):
    return (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)

for y in [1900, 2000, 2024, 2025]:
    print(f"{y}: {'Leap' if is_leap(y) else 'Not leap'}")
```

### Challenge 4: Traffic Light Simulator
Given a colour string, print what the driver should do.

```python
def traffic_action(colour):
    match colour.lower():
        case "red":
            return "Stop"
        case "yellow" | "amber":
            return "Slow down"
        case "green":
            return "Go"
        case _:
            return "Unknown signal"

for c in ["red", "green", "amber", "blue"]:
    print(f"{c}: {traffic_action(c)}")
```

### Challenge 5: Simple Login System
Store a username and password. Give the user 3 attempts before locking out.

```python
CORRECT_USER = "admin"
CORRECT_PASS = "python123"
MAX_ATTEMPTS = 3

for attempt in range(1, MAX_ATTEMPTS + 1):
    username = input("Username: ")
    password = input("Password: ")

    if username == CORRECT_USER and password == CORRECT_PASS:
        print("Login successful!")
        break
    else:
        remaining = MAX_ATTEMPTS - attempt
        if remaining > 0:
            print(f"Incorrect. {remaining} attempt(s) left.")
        else:
            print("Account locked. Too many failed attempts.")
```
