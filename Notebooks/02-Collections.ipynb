#### 🐍 Python Collections Module

Each topic includes Definition, Use Cases, and Example.

## 1. namedtuple

**Definition:** A tuple subclass that allows fields to be accessed by name instead of index.

**Common Use Cases:**
- Representing coordinates
- Database records
- Configuration objects


```python
from collections import namedtuple

Employee = namedtuple('Employee', ['id','name','dept'])
emp = Employee(101,'Amit','IT')
print(emp)
print(emp.name)
```

## 2. deque

**Definition:** A double-ended queue that supports fast insertion and deletion from both ends.

**Common Use Cases:**
- Task queues
- Browser history
- Sliding window problems


```python
from collections import deque

dq = deque([1,2,3])
dq.appendleft(0)
dq.append(4)
print(dq)
dq.popleft()
print(dq)
```

## 3. Counter

**Definition:** A dictionary subclass used to count the frequency of hashable objects.

**Common Use Cases:**
- Word frequency analysis
- Log analysis
- Vote counting


```python
from collections import Counter

text='banana'
count=Counter(text)
print(count)
print(count.most_common(2))
```

## 4. OrderedDict

**Definition:** A dictionary that preserves insertion order.

**Common Use Cases:**
- LRU Cache implementations
- Ordered reports
- Configuration settings


```python
from collections import OrderedDict

od=OrderedDict()
od['SQL']=1
od['Python']=2
od['PySpark']=3
print(od)
```

## 5. defaultdict

**Definition:** A dictionary that automatically provides a default value for missing keys.

**Common Use Cases:**
- Grouping records
- Counting items
- Building adjacency lists for graphs


```python
from collections import defaultdict

dd=defaultdict(list)
dd['IT'].append('Amit')
dd['IT'].append('Neha')
dd['HR'].append('Raj')
print(dd)
```
