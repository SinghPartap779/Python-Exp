# Python Variables & Storage Cheat Sheet

## 1. Variable Assignment
Variables are created on assignment — no declaration needed. Python infers the type.

```python
name   = "Alice"     # str
age    = 25          # int
height = 5.6         # float
is_student = True    # bool

print(name, age, height, is_student)
print(type(name), type(age), type(height), type(is_student))
```

## 2. Multiple & Chained Assignment
Assign to many variables in one line.

```python
# Assign different values at once
x, y, z = 1, 2, 3
print(x, y, z)   # 1 2 3

# Assign the same value to multiple variables
a = b = c = 0
print(a, b, c)   # 0 0 0

# Swap without a temp variable
a, b = 10, 20
a, b = b, a
print(a, b)      # 20 10
```

## 3. Data Types Overview
Python has several built-in types — all are objects.

```python
i   = 42            # int
f   = 3.14          # float
c   = 2 + 3j        # complex
s   = "hello"       # str
b   = True          # bool
n   = None          # NoneType
lst = [1, 2, 3]     # list
tup = (1, 2, 3)     # tuple
dct = {"k": "v"}    # dict
st  = {1, 2, 3}     # set

for var in [i, f, c, s, b, n, lst, tup, dct, st]:
    print(f"{str(var):<20} -> {type(var).__name__}")
```

## 4. Type Checking & Conversion
`type()`, `isinstance()`, and casting functions.

```python
x = "123"

# Check type
print(type(x))            # <class "str">
print(type(x) == str)     # True
print(isinstance(x, str)) # True  (preferred — works with inheritance)

# Type conversion (casting)
print(int("42"))          # 42
print(float("3.14"))      # 3.14
print(str(100))           # "100"
print(bool(0))            # False
print(bool("hello"))      # True  (non-empty string is truthy)
print(list((1, 2, 3)))    # [1, 2, 3]
print(tuple([4, 5, 6]))   # (4, 5, 6)
```

## 5. Dynamic Typing
A variable can point to a different type at any time.

```python
var = 10
print(var, type(var))   # 10 <class "int">

var = "hello"
print(var, type(var))   # hello <class "str">

var = [1, 2, 3]
print(var, type(var))   # [1, 2, 3] <class "list">

# This is fine in Python — the variable is just a label
# pointing to an object in memory.
```

## 6. Variable Naming Rules
Names can use letters, digits, and underscores — but cannot start with a digit.

```python
# Valid names
my_var   = 1
_private = 2
camelCase = 3
var99    = 4

# Python naming conventions
snake_case_variable = "preferred for variables and functions"
CONSTANT_VALUE      = 3.14159   # ALL_CAPS for constants (convention)
ClassName           = "PascalCase for classes"

# Reserved keywords — cannot be used as variable names
import keyword
print(keyword.kwlist)
```

## 7. Constants
Python has no built-in `const`. By convention, ALL_CAPS signals a constant.

```python
PI       = 3.14159
MAX_SIZE = 100
APP_NAME = "MyApp"

print(PI, MAX_SIZE, APP_NAME)

# For true immutability, use a named tuple or dataclass with frozen=True
from collections import namedtuple
Config = namedtuple("Config", ["host", "port"])
config = Config(host="localhost", port=8080)
print(config.host, config.port)
# config.host = "other"  ->  AttributeError
```

## 8. Memory & Identity
`id()` returns the memory address. `is` checks identity (same object), `==` checks equality (same value).

```python
a = [1, 2, 3]
b = [1, 2, 3]
c = a

print(a == b)    # True  — same values
print(a is b)    # False — different objects in memory
print(a is c)    # True  — c points to the same object as a

print(id(a), id(b), id(c))

# Small integers and interned strings are cached by Python
x = 5
y = 5
print(x is y)    # True  — Python reuses small int objects (-5 to 256)
```

## 9. Variable Scope
LEGB rule: **L**ocal → **E**nclosing → **G**lobal → **B**uilt-in

```python
x = "global"

def outer():
    x = "enclosing"
    def inner():
        x = "local"
        print("inner :", x)   # local
    inner()
    print("outer :", x)       # enclosing

outer()
print("global:", x)           # global
```

```python
count = 0

def increment():
    global count        # declare intent to modify global
    count += 1

increment()
increment()
print(count)  # 2

def make_counter():
    n = 0
    def step():
        nonlocal n      # modify the enclosing scope variable
        n += 1
        return n
    return step

counter = make_counter()
print(counter())  # 1
print(counter())  # 2
```

## 10. Deleting Variables & Type Hints
`del` removes a name binding. Type hints document intent without enforcing types.

```python
# Delete a variable
temp = 42
print(temp)   # 42
del temp
# print(temp) ->  NameError: name "temp" is not defined

# Type hints (Python 3.5+) — purely informational
name: str = "Alice"
age:  int = 25
pi:   float = 3.14

def greet(person: str, times: int = 1) -> str:
    return f"Hello, {person}! " * times

print(greet("Bob", 2))
```

Print the type and value of each item in `[42, 3.14, "hello", True, None]`.

### Challenge 2: Type Conversion Chain
Start with the string `"9"`, convert it to `int`, then to `float`, then back to `str` and print each step.

```python
s = "9"
i = int(s)
f = float(i)
back = str(f)
print(s, type(s))      # 9     <class str>
print(i, type(i))      # 9     <class int>
print(f, type(f))      # 9.0   <class float>
print(back, type(back))# 9.0   <class str>
```

### Challenge 3: Scope Experiment
Create a global variable `total = 0`, write a function that adds a number to it using `global`, call it three times.

```python
total = 0

def add(n):
    global total
    total += n

add(10)
add(25)
add(5)
print(total)  # 40
```

### Challenge 4: Identity vs Equality
Create two lists with the same values. Show that `==` is `True` but `is` is `False`.

```python
list1 = [1, 2, 3]
list2 = [1, 2, 3]

print("Equal?   ", list1 == list2)  # True
print("Same obj?", list1 is list2)  # False
print("id list1:", id(list1))
print("id list2:", id(list2))
```
