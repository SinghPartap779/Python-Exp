# Python Tuple Operations Cheat Sheet

## 1. Creating Tuples
Tuples are **ordered**, **immutable** sequences. Use `()` or just commas.

```python
t1 = (1, 2, 3)
t2 = 1, 2, 3           # parentheses are optional
t3 = tuple([4, 5, 6])  # from a list
t4 = tuple("hello")    # from a string

# Single-element tuple MUST have a trailing comma
single = (42,)          # tuple
not_tuple = (42)        # just an int!

empty = ()

print(t1)       # (1, 2, 3)
print(t4)       # ("h", "e", "l", "l", "o")
print(type(single), type(not_tuple))  # <class "tuple"> <class "int">
```

## 2. Accessing & Slicing
Same indexing rules as lists — indexing starts at 0.

```python
t = ("a", "b", "c", "d", "e")

print(t[0])     # a      (first element)
print(t[-1])    # e      (last element)
print(t[1:3])   # ("b", "c")  (slice)
print(t[::2])   # ("a", "c", "e")  (every other)
print(t[::-1])  # ("e", "d", "c", "b", "a")  (reversed)
print(len(t))   # 5
```

## 3. Tuple Unpacking
Assign tuple elements directly to variables — including the `*` wildcard.

```python
# Basic unpacking
point = (10, 20)
x, y = point
print(x, y)   # 10 20

# Swap two variables without a temp variable
a, b = 5, 9
a, b = b, a
print(a, b)   # 9 5

# Extended unpacking with *
first, *middle, last = (1, 2, 3, 4, 5)
print(first)   # 1
print(middle)  # [2, 3, 4]
print(last)    # 5

# Ignore values with _
name, _, score = ("Alice", "ID-001", 95)
print(name, score)  # Alice 95
```

## 4. Tuple Methods
Tuples only have two built-in methods: `count()` and `index()`.

```python
t = (1, 2, 3, 2, 4, 2, 5)

# count() — how many times a value appears
print(t.count(2))      # 3
print(t.count(9))      # 0  (no error)

# index() — position of first occurrence
print(t.index(3))      # 2
print(t.index(2))      # 1  (first 2)

# index() with start/end bounds
print(t.index(2, 2))   # 3  (search from position 2 onwards)

# t.index(99)  ->  ValueError if not found
```

## 5. Immutability
Tuples cannot be changed after creation — but a tuple can *contain* mutable objects.

```python
t = (1, 2, 3)
# t[0] = 99  ->  TypeError: "tuple" object does not support item assignment

# Workaround: convert to list, modify, convert back
lst = list(t)
lst[0] = 99
t = tuple(lst)
print(t)   # (99, 2, 3)

# A tuple CAN hold a mutable object like a list
mixed = (1, [2, 3], 4)
mixed[1].append(99)       # the list inside is still mutable
print(mixed)  # (1, [2, 3, 99], 4)
```

## 6. Iterating & Membership
`for` loops, `in`, `enumerate()`, `zip()`

```python
colors = ("red", "green", "blue")

# Basic iteration
for color in colors:
    print(color)
print()

# With index using enumerate()
for i, color in enumerate(colors):
    print(i, color)
print()

# Membership test
print("red" in colors)    # True
print("yellow" in colors) # False

# Zip two tuples together
names  = ("Alice", "Bob", "Carol")
scores = (85, 92, 78)
for name, score in zip(names, scores):
    print(f"{name}: {score}")
```

## 7. Tuple as Dictionary Key & in Sets
Because tuples are immutable and hashable they can be used where lists cannot.

```python
# Tuple as a dictionary key (lists cannot be keys)
location = {}
location[(28.6, 77.2)] = "Delhi"
location[(19.0, 72.8)] = "Mumbai"
print(location[(28.6, 77.2)])  # Delhi

# Tuples inside a set
visited = {(28.6, 77.2), (19.0, 72.8), (28.6, 77.2)}
print(visited)  # only unique coordinates kept
print(len(visited))  # 2
```

## 8. Nested Tuples
Tuples can contain other tuples — useful for representing tables or matrices.

```python
matrix = (
    (1, 2, 3),
    (4, 5, 6),
    (7, 8, 9),
)

print(matrix[1])       # (4, 5, 6)  — second row
print(matrix[1][2])    # 6           — row 1, col 2

# Iterate nested tuples
for row in matrix:
    print(row)
```

## 9. Useful Built-in Functions with Tuples
`sorted()`, `min()`, `max()`, `sum()`, `any()`, `all()`

```python
t = (4, 1, 7, 2, 9, 3)

print(min(t))       # 1
print(max(t))       # 9
print(sum(t))       # 26

# sorted() returns a LIST (tuples are immutable)
print(sorted(t))            # [1, 2, 3, 4, 7, 9]
print(sorted(t, reverse=True))  # [9, 7, 4, 3, 2, 1]

flags = (True, True, False)
print(any(flags))   # True  — at least one is True
print(all(flags))   # False — not all are True
```

## 10. Named Tuples
`collections.namedtuple` gives tuple fields readable names — like a lightweight class.

```python
from collections import namedtuple

Point = namedtuple("Point", ["x", "y"])
p = Point(3, 7)

print(p)       # Point(x=3, y=7)
print(p.x)     # 3  (by name)
print(p[0])    # 3  (by index — still a tuple)

Student = namedtuple("Student", ["name", "age", "grade"])
alice = Student("Alice", 20, "A")
print(f"{alice.name} got grade {alice.grade}")

# _replace() returns a new tuple with one field changed
alice_senior = alice._replace(age=21)
print(alice_senior)
```

### Challenge 2: Unpack a Record

```python
record = ("Alice", 25, "Engineer")
name, age, job = record
print(f"{name} is {age} years old and works as an {job}.")
```

### Challenge 3: Most Frequent Element
Find the most frequent number in `(3, 1, 4, 1, 5, 9, 2, 6, 1, 3)`.

```python
t = (3, 1, 4, 1, 5, 9, 2, 6, 1, 3)
most_frequent = max(set(t), key=t.count)
print(f"Most frequent: {most_frequent} (appears {t.count(most_frequent)} times)")  # 1 (3 times)
```

### Challenge 4: Coordinate Store
Store city names keyed by `(latitude, longitude)` tuples and look one up.

```python
cities = {
    (28.6, 77.2): "Delhi",
    (19.0, 72.8): "Mumbai",
    (12.9, 77.6): "Bengaluru",
}
coords = (19.0, 72.8)
print(f"City at {coords}: {cities[coords]}")  # Mumbai
```
