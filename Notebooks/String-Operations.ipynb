# Python String Operations Cheat Sheet

## 1. Creating & Basic Info
Strings can be single `'...'`, double `"..."`, or triple quotes `'''...'''`.

```python
s = "Hello World"
print(len(s))  # 11
```

## 2. Accessing & Slicing
Indexing starts at 0.

```python
s = "Python"
print(s[0])     # P
print(s[-1])    # n
print(s[0:3])   # Pyt
print(s[::2])   # Pto
```

## 3. Case Conversion
`upper()`, `lower()`, `title()`, `capitalize()`, `swapcase()`

```python
s = "hello world"
print(s.upper())      # HELLO WORLD
print(s.title())      # Hello World
print(s.capitalize()) # Hello world
print(s.swapcase())   # HELLO WORLD -> hello world
```

## 4. Searching & Checking
`find()`, `index()`, `startswith()`, `endswith()`, `in`

```python
s = "Python Programming"
print(s.find("Pro"))       # 7
print(s.startswith("Py"))  # True
print("gram" in s)         # True
```

## 5. Replacing & Splitting
`replace()`, `split()`, `join()`

```python
s = "I love Python"
print(s.replace("love", "like"))  # I like Python

words = s.split(" ")
print(words)  # ['I', 'love', 'Python']

joined = "-".join(words)
print(joined) # I-love-Python
```

## 6. Stripping & Padding
`strip()`, `lstrip()`, `rstrip()`, `center()`, `zfill()`

```python
s = "   hello   "
print(s.strip())   # hello

s = "42"
print(s.zfill(5))  # 00042

s = "Python"
print(s.center(10, "*"))  # **Python**
```

## 7. Validation Methods
`isalnum()`, `isalpha()`, `isdigit()`, `isspace()`

```python
print("123".isdigit())    # True
print("abc".isalpha())    # True
print("abc123".isalnum()) # True
print("   ".isspace())    # True
```

## 8. String Formatting
f-strings, `format()`

```python
name = "Alice"
age = 25
print(f"My name is {name}, I am {age} years old.")
print("My name is {}, I am {} years old.".format(name, age))
```

## Practice Challenges

### Challenge 1: Palindrome Check
Write a program to check if a string is a palindrome.

```python
def is_palindrome(s):
    s = s.lower().replace(" ", "")
    return s == s[::-1]

print(is_palindrome("racecar"))  # True
print(is_palindrome("hello"))    # False
print(is_palindrome("A man a plan a canal Panama"))  # True
```

### Challenge 2: Count Character Occurrences
Count how many times "a" appears in "banana".

```python
word = "banana"
count = word.count("a")
print(f'"a" appears {count} times in "{word}"')  # 3 times
```
