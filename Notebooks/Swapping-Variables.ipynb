# Python Swapping Variables Cheat Sheet

## 1. Swap Using Tuple Unpacking (Pythonic Way)
The cleanest and most readable method — Python evaluates the right side first.

```python
a = 10
b = 20

a, b = b, a

print(a)  # 20
print(b)  # 10
```

## 2. Swap Using a Temporary Variable
The classic approach used in most other languages.

```python
a = 100
b = 200

temp = a
a = b
b = temp

print(a)  # 200
print(b)  # 100
```

## 3. Swap Using Arithmetic (Add & Subtract)
Works for numbers only — no extra variable needed, but less readable.

```python
a = 5
b = 3

a = a + b   # a = 8
b = a - b   # b = 5  (original a)
a = a - b   # a = 3  (original b)

print(a)  # 3
print(b)  # 5

# Limitation: can overflow with very large integers in other languages
# In Python integers are unbounded, so it always works for ints
```

## 4. Swap Using XOR (Bitwise)
Works for integers only — a classic interview trick.

```python
a = 7   # binary: 0111
b = 4   # binary: 0100

a = a ^ b   # a = 0011  (3)
b = a ^ b   # b = 0111  (7, original a)
a = a ^ b   # a = 0100  (4, original b)

print(a)  # 4
print(b)  # 7

# WARNING: breaks if a and b point to the same object (a is b)
# Always prefer tuple unpacking in Python
```

## 5. Swapping Multiple Variables at Once
Tuple unpacking scales to any number of variables in one line.

```python
a, b, c = 1, 2, 3
print(a, b, c)   # 1 2 3

# Rotate left: a <- b <- c <- a
a, b, c = b, c, a
print(a, b, c)   # 2 3 1

# Rotate right
a, b, c = c, a, b
print(a, b, c)   # 1 2 3  (back to original)
```

## 6. Swapping Elements in a List
Use index-based tuple unpacking to swap any two positions.

```python
nums = [10, 20, 30, 40, 50]
print("Before:", nums)

# Swap index 1 and index 3
nums[1], nums[3] = nums[3], nums[1]
print("After :", nums)  # [10, 40, 30, 20, 50]

# Swap first and last
nums[0], nums[-1] = nums[-1], nums[0]
print("After :", nums)  # [50, 40, 30, 20, 10]
```

## 7. Swapping Inside a Function
Functions return multiple values as tuples — perfect for swapping.

```python
def swap(x, y):
    return y, x

a, b = 5, 9
a, b = swap(a, b)
print(a, b)  # 9 5

# Generic swap for any two values
def swap_and_show(x, y):
    print(f"Before: x={x}, y={y}")
    x, y = y, x
    print(f"After : x={x}, y={y}")
    return x, y

swap_and_show("hello", "world")
```

## 8. Swapping Dictionary Values
Swap the values associated with two keys.

```python
data = {"first": "Alice", "second": "Bob"}
print("Before:", data)

data["first"], data["second"] = data["second"], data["first"]
print("After :", data)  # {"first": "Bob", "second": "Alice"}

# Swap keys mapping to specific values
scores = {"Math": 90, "Science": 85}
scores["Math"], scores["Science"] = scores["Science"], scores["Math"]
print(scores)  # {"Math": 85, "Science": 90}
```

## 9. Swapping in Loops — Bubble Sort
Swapping is the core operation in sorting algorithms like Bubble Sort.

```python
nums = [64, 34, 25, 12, 22, 11, 90]
n = len(nums)

for i in range(n - 1):
    for j in range(n - 1 - i):
        if nums[j] > nums[j + 1]:
            nums[j], nums[j + 1] = nums[j + 1], nums[j]  # swap

print("Sorted:", nums)  # [11, 12, 22, 25, 34, 64, 90]
```

## 10. Swapping with String & Type Comparison
Always ensure types are compatible; swapping works on any objects.

```python
# Strings
first_name = "John"
last_name  = "Doe"
first_name, last_name = last_name, first_name
print(first_name, last_name)  # Doe John

# Mixed types — Python does not care
x = 42
y = "hello"
x, y = y, x
print(x, type(x))  # hello <class str>
print(y, type(y))  # 42    <class int>

# Nested swap — swap rows in a 2D list
matrix = [[1, 2], [3, 4], [5, 6]]
matrix[0], matrix[2] = matrix[2], matrix[0]
print(matrix)  # [[5, 6], [3, 4], [1, 2]]
```

### Challenge 5: Sort Two Numbers Without sort()
Given `a = 50` and `b = 20`, ensure `a` always holds the smaller value using a conditional swap.

```python
a = 50
b = 20

if a > b:
    a, b = b, a

print(f"Smaller: {a}, Larger: {b}")  # Smaller: 20, Larger: 50
```
