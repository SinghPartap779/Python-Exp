## Importing the math Module

The `math` module is a built-in Python library that provides mathematical functions and constants.


```python
import math

print(math.pi)
print(math.e)
```

## 1. Mathematical Constants

### Definitions
- `math.pi` → Value of π
- `math.e` → Euler's Number
- `math.tau` → 2π
- `math.inf` → Infinity
- `math.nan` → Not a Number

### Use Cases
- Geometry calculations
- Scientific calculations
- Statistical applications


```python
import math

print("PI:", math.pi)
print("Euler:", math.e)
print("Tau:", math.tau)
print("Infinity:", math.inf)
print("NaN:", math.nan)
```

## 2. Number Theoretic Functions

### Definitions
- `factorial()` → Product of positive integers
- `gcd()` → Greatest Common Divisor
- `lcm()` → Least Common Multiple
- `isqrt()` → Integer square root

### Use Cases
- Combinatorics
- Cryptography
- Mathematics


```python
import math

print(math.factorial(5))
print(math.gcd(24, 36))
print(math.lcm(12, 18))
print(math.isqrt(25))
```

## 3. Power, Exponential and Logarithmic Functions

### Use Cases
- Growth calculations
- Financial modeling
- Machine Learning


```python
import math

print(math.sqrt(25))
print(math.pow(2, 5))
print(math.exp(2))
print(math.log(10))
print(math.log10(100))
print(math.log2(8))
```

## 4. Trigonometric Functions

### Use Cases
- Engineering
- Physics
- Graphics Programming
- Navigation


```python
import math

angle = math.radians(30)

print(math.sin(angle))
print(math.cos(angle))
print(math.tan(angle))

print(math.degrees(angle))
```

## 5. Rounding Functions

### Use Cases
- Financial calculations
- Report generation
- Data cleaning


```python
import math

x = 5.89

print(math.ceil(x))
print(math.floor(x))
print(math.trunc(x))
print(math.fabs(-25.7))
```

## 6. Combinations and Permutations

### Use Cases
- Probability
- Statistics
- Interview Questions


```python
import math

print("Combination:", math.comb(5, 2))
print("Permutation:", math.perm(5, 2))
```

## 7. Distance and Norm Functions

### Use Cases
- Machine Learning
- GIS
- Coordinate Geometry


```python
import math

print(math.dist((1, 2), (4, 6)))
print(math.hypot(3, 4))
```

## Real World Example - Area of Circle

Formula:
Area = πr²


```python
import math

radius = 5
area = math.pi * (radius ** 2)

print("Area of Circle =", area)
```
