## 2. Number Theory
`factorial(n)` ┬╖ `gcd(a,b)` ┬╖ `lcm(a,b)` ┬╖ `isqrt(n)`
**Use:** combinatorics ┬╖ cryptography ┬╖ simplifying fractions

```python
import math

# How many ways to arrange 5 books? тЖТ 5! = 120
print("5! =", math.factorial(5))

# Simplify fraction 24/36 тЖТ divide both by GCD
print("GCD(24,36) =", math.gcd(24, 36))       # 12 тЖТ 24/12=2, 36/12=3

# LCM: next meeting time if two buses run every 12 and 18 mins
print("LCM(12,18) =", math.lcm(12, 18))       # 36 minutes

# Integer square root (floor) тАФ no floats
print("isqrt(30) =", math.isqrt(30))          # 5  (not 5.47...)
```

## 3. Power, Exponential & Logarithmic
`sqrt` ┬╖ `pow` ┬╖ `exp` ┬╖ `log` ┬╖ `log10` ┬╖ `log2`
**Use:** compound interest ┬╖ ML loss functions ┬╖ decibel calculations

```python
import math

print("sqrt(144)  :", math.sqrt(144))          # 12.0
print("pow(2, 10) :", math.pow(2, 10))         # 1024.0  (returns float)
print("2**10      :", 2 ** 10)                 # 1024    (returns int)

# Compound interest: how many years to double at 8%?
rate = 0.08
years = math.log(2) / math.log(1 + rate)
print(f"\nYears to double at 8%: {years:.1f} years")  # Rule of 72 тЙИ 9

print("\nlog(e)    :", math.log(math.e))       # 1.0  (natural log)
print("log10(100):", math.log10(100))          # 2.0
print("log2(8)   :", math.log2(8))             # 3.0
```

## 4. Trigonometric Functions
`sin` ┬╖ `cos` ┬╖ `tan` and their inverses `asin` ┬╖ `acos` ┬╖ `atan`
**Note:** functions expect **radians** тАФ use `math.radians()` to convert from degrees
**Use:** engineering ┬╖ physics ┬╖ graphics ┬╖ navigation

```python
import math

# Always convert to radians first
angle_deg = 30
angle_rad = math.radians(angle_deg)

print(f"sin({angle_deg}┬░) = {math.sin(angle_rad):.4f}")   # 0.5
print(f"cos({angle_deg}┬░) = {math.cos(angle_rad):.4f}")   # 0.866
print(f"tan({angle_deg}┬░) = {math.tan(angle_rad):.4f}")   # 0.577

# Inverse: find the angle from a ratio
ratio = 0.5
angle_back = math.degrees(math.asin(ratio))
print(f"\nasin(0.5) = {angle_back}┬░")          # 30.0┬░
```

## 5. Rounding
| Function | Behaviour | Example |
|----------|-----------|---------|
| `ceil(x)` | Round **up** | ceil(5.1) тЖТ 6 |
| `floor(x)` | Round **down** | floor(5.9) тЖТ 5 |
| `trunc(x)` | Drop decimal | trunc(-5.9) тЖТ -5 |
| `fabs(x)` | Absolute value (float) | fabs(-25.7) тЖТ 25.7 |

```python
import math

x = 5.89

print(math.ceil(x))
print(math.floor(x))
print(math.trunc(x))
print(math.fabs(-25.7))
```

## 6. Combinations & Permutations
`comb(n, r)` = C(n,r) тАФ order does NOT matter
`perm(n, r)` = P(n,r) тАФ order DOES matter
**Use:** probability ┬╖ statistics ┬╖ lottery problems

```python
import math

print("Combination:", math.comb(5, 2))
print("Permutation:", math.perm(5, 2))
```

## 7. Distance Functions
`dist(p, q)` тАФ Euclidean distance between two points
`hypot(x, y)` тАФ length of hypotenuse (equivalent to `sqrt(x┬▓+y┬▓)`)
**Use:** ML (nearest neighbour) ┬╖ GIS ┬╖ geometry

```python
import math

# Distance between two GPS coordinates (simplified)
point_a = (0, 0)
point_b = (3, 4)
print("Distance:", math.dist(point_a, point_b))   # 5.0 (3-4-5 triangle)

# Hypotenuse тАФ same result
print("Hypotenuse:", math.hypot(3, 4))             # 5.0

# 3D distance
p1 = (1, 2, 3)
p2 = (4, 6, 3)
print("3D Distance:", math.dist(p1, p2))           # 5.0
```

## Real-World Example тАФ Circle Area
`Area = ╧А ├Ч r┬▓`

```python
import math

radius = 5
area = math.pi * (radius ** 2)

print("Area of Circle =", area)
```
