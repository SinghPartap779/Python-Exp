# 1. What is a Loop?

A loop is used to execute a block of code repeatedly.

Without loops:

```python
print("Hello")
print("Hello")
print("Hello")
```

With loops:

```python
for i in range(3):
    print("Hello")
```

## Use Cases

- Reading files
- Processing records
- ETL pipelines
- API pagination
- Data analysis
- Automation scripts


# 2. for Loop

````python
for i in range(5):
    print(i)
````


## How it Works

range(5) generates:

0, 1, 2, 3, 4

Loop executes once for each value.


# 3. range() Function

````python
print(list(range(5)))
print(list(range(2,10)))
print(list(range(0,20,2)))
````


### Syntax

# 6. While Loop

````python
count = 1

while count <= 5:
    print(count)
    count += 1
````


## When to Use While Loop?

Use when number of iterations is unknown.

# 8. break Statement

````python
for i in range(10):

    if i == 5:
        break

    print(i)
````


### Use Case

Stop processing when record found.


# 9. continue Statement

````python
for i in range(10):

    if i == 5:
        continue

    print(i)
````


### Use Case

Skip invalid records.


# 10. pass Statement

````python
for i in range(5):

    if i == 3:
        pass

    print(i)
````

# 11. Nested Loops

````python
for row in range(3):

    for col in range(3):
        print(row, col)
````


### Use Cases

- Matrix operations
- Chess boards
- Seating arrangements


# 12. Pattern Programs

````python
for i in range(5):
    print("*" * i)
````

# 13. Loop Else

````python
for i in range(5):
    print(i)
else:
    print("Loop Completed")
````


The else block executes when loop finishes normally.


# 14. enumerate()

````python
names = ["John","David","Mike"]

for index, value in enumerate(names):
    print(index, value)
````

# 15. zip()

````python
names = ["John","Mike"]
scores = [90,95]

for n,s in zip(names,scores):
    print(n,s)
````

# 16. Real World ETL Example

````python
sales = [100,200,300,400]

total = 0

for amount in sales:
    total += amount
    #total = total + amount

print("Total Sales =", total)
````
