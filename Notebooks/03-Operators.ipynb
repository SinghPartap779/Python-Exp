#### Python Operators
This notebook explains Python operators with:
- Definitions
- Syntax
- Examples
- Real-world use cases

## Arithmetic Operators

Arithmetic operators perform mathematical calculations.

**Use Cases**
- Salary calculations
- Percentage calculations
- Scientific computations
- Financial applications


```python
a = 10
b = 3

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Exponent:", a ** b)
```

## Comparison Operators

Comparison operators compare two values and return True or False.

**Use Cases**
- Age validation
- Eligibility checks
- Filtering records
- Decision making


```python
a = 10
b = 20

print(a == b)
print(a != b)
print(a > b)
print(a < b)
print(a >= b)
print(a <= b)
```

## Logical Operators

Logical operators combine multiple conditions.

**Use Cases**
- Login systems
- Access control
- Business rules
- Data validation


```python
x = True
y = False

print(x and y)
print(x or y)
print(not x)
```

## Assignment Operators

Assignment operators assign and update values.

**Use Cases**
- Counters
- Running totals
- Aggregations


```python
x = 10
x += 5
print(x)

x *= 2
print(x)
```

## Identity Operators

Identity operators check whether two variables refer to the same object in memory.

**Use Cases**
- Object comparison
- Memory/reference checks


```python
a = [1,2,3]
b = a
c = [1,2,3]

print(a is b)
print(a is c)
print(a == c)
```

## Membership Operators

Membership operators check whether a value exists in a sequence.

**Use Cases**
- Searching data
- Validation
- Access control


```python
nums = [1,2,3,4]

print(2 in nums)
print(5 not in nums)
```

## Bitwise Operators

Bitwise operators work on binary representations.

**Use Cases**
- Networking
- Encryption
- Embedded systems
- Performance optimization


```python
a = 10
b = 4

print(a & b)
print(a | b)
print(a ^ b)
print(~a)
print(a << 2)
print(a >> 2)
```

# Operator Precedence

Python follows precedence rules:

1. Parentheses ()
2. Exponentiation **
3. Multiplication, Division, Floor Division, Modulus
4. Addition, Subtraction
5. Comparison Operators
6. Logical Operators

Example:


```python
result = 10 + 5 * 2
print(result)  # 20
```
