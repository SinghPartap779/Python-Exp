# Python Comprehensions

Comprehensions are a concise, readable way to create **lists, dicts, sets, and generators** from existing iterables.

| Type | Syntax | Returns |
|------|--------|---------|
| List | `[expr for item in iterable if cond]` | `list` |
| Dict | `{k: v for item in iterable if cond}` | `dict` |
| Set  | `{expr for item in iterable if cond}` | `set` |
| Generator | `(expr for item in iterable if cond)` | `generator` (lazy) |

## 1. List Comprehension

**Pattern:** `[expression for item in iterable]`  
Replaces the common `for` loop + `.append()` pattern in one readable line.

````python
# Traditional loop vs list comprehension
numbers = [1, 2, 3, 4, 5]

# Old way
squares_loop = []
for n in numbers:
    squares_loop.append(n ** 2)

# List comprehension — same result, one line
squares = [n ** 2 for n in numbers]

print("Loop result  :", squares_loop)   # [1, 4, 9, 16, 25]
print("Comprehension:", squares)        # [1, 4, 9, 16, 25]

# Works on any iterable
upper_names = [name.upper() for name in ["alice", "bob", "charlie"]]
print("Upper names  :", upper_names)    # ['ALICE', 'BOB', 'CHARLIE']

# Range as source
evens = [n for n in range(1, 11) if n % 2 == 0]
print("Evens 1-10   :", evens)          # [2, 4, 6, 8, 10]
````

## 2. List Comprehension with `if` Filter

Add a condition at the end to **filter** which items are included.

````python
data = [10, -3, 45, -8, 22, -1, 0, 7]

# Only positive numbers
positives = [x for x in data if x > 0]
print("Positives :", positives)          # [10, 45, 22, 7]

# Square only the positives
pos_squares = [x ** 2 for x in data if x > 0]
print("Pos squares:", pos_squares)       # [100, 2025, 484, 49]

# Filter strings by length
words = ["cat", "elephant", "dog", "hippopotamus", "ox"]
long_words = [w for w in words if len(w) > 4]
print("Long words :", long_words)        # ['elephant', 'hippopotamus']

# if-else inside expression (ternary) — different from filter!
# This applies to EVERY item; it changes the value, not filters items
labeled = ["pos" if x > 0 else "neg" for x in data]
print("Labeled    :", labeled)
````

## 3. Nested List Comprehension

Flatten a 2D list or create a matrix — read the `for` clauses **left to right** (outer loop first).

````python
# Flatten a 2D list
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

flat = [num for row in matrix for num in row]
print("Flattened:", flat)   # [1, 2, 3, 4, 5, 6, 7, 8, 9]

# Build a multiplication table (list of lists)
table = [[row * col for col in range(1, 4)] for row in range(1, 4)]
print("\nMultiplication table:")
for row in table:
    print(" ", row)

# Pairs — all combinations of two lists
colors  = ["red", "blue"]
sizes   = ["S", "M", "L"]
combos  = [(c, s) for c in colors for s in sizes]
print("\nColor-Size combos:", combos)
````

## 4. Dictionary Comprehension

**Pattern:** `{key_expr: value_expr for item in iterable}`  
Build a dict from any iterable in one line.

````python
# Square each number as key → value
sq_map = {n: n ** 2 for n in range(1, 6)}
print("Squares map:", sq_map)   # {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}

# Word → length mapping
words = ["apple", "banana", "cherry"]
lengths = {w: len(w) for w in words}
print("Word lengths:", lengths)

# Invert a dictionary (swap keys and values)
original = {"a": 1, "b": 2, "c": 3}
inverted = {v: k for k, v in original.items()}
print("Inverted    :", inverted)   # {1: 'a', 2: 'b', 3: 'c'}

# Filter: keep only passing scores
scores = {"Alice": 85, "Bob": 42, "Charlie": 91, "Diana": 55}
passed = {name: score for name, score in scores.items() if score >= 60}
print("Passed      :", passed)

# Uppercase keys
up_keys = {k.upper(): v for k, v in scores.items()}
print("Upper keys  :", up_keys)
````

## 5. Set Comprehension

**Pattern:** `{expr for item in iterable}`  
Same as list comprehension but uses `{}` — result is a **set** (unique, unordered).

````python
# Unique lengths from a list of words
words = ["cat", "dog", "elephant", "ant", "bear", "emu"]
unique_lengths = {len(w) for w in words}
print("Unique lengths:", unique_lengths)   # {3, 8} — order not guaranteed

# Remove duplicates from a list while transforming
data = [1, 2, 2, 3, 3, 3, 4]
unique_squares = {n ** 2 for n in data}
print("Unique squares:", unique_squares)   # {1, 4, 9, 16}

# Unique first characters
names = ["Alice", "Bob", "Anna", "Charlie", "Brian"]
first_chars = {name[0] for name in names}
print("First chars   :", first_chars)      # {'A', 'B', 'C'}
````

## 6. Generator Expression

**Pattern:** `(expr for item in iterable)`  
Looks like a list comprehension with `()` instead of `[]`.  
Key difference: it is **lazy** — values are produced one at a time, never all stored in memory.

| | List `[]` | Generator `()` |
|-|-----------|----------------|
| Memory | All at once | One at a time |
| Reusable | Yes | **No** (exhausted after one pass) |
| Speed | Faster for small data | Faster for large data |

````python
import sys

# List vs generator — same values, very different memory usage
lst = [n ** 2 for n in range(10_000)]
gen = (n ** 2 for n in range(10_000))

print("List size     :", sys.getsizeof(lst), "bytes")
print("Generator size:", sys.getsizeof(gen), "bytes")   # ~120 bytes regardless of size!

# Generators work great with sum(), max(), min() — no intermediate list needed
total = sum(n ** 2 for n in range(1, 101))
print("\nSum of squares 1-100:", total)

# Iterating a generator — once consumed, it's empty
gen2 = (x * 2 for x in range(5))
print("\nFirst pass :", list(gen2))   # [0, 2, 4, 6, 8]
print("Second pass:", list(gen2))   # []  ← exhausted!

# next() pulls one value at a time
gen3 = (x ** 2 for x in range(5))
print("\nnext() calls:")
print(next(gen3))   # 0
print(next(gen3))   # 1
print(next(gen3))   # 4
````

## 7. Real-World Examples

Comprehensions are used constantly in data processing — especially when cleaning or transforming datasets before loading into PySpark.

````python
# --- CSV-style raw data (list of strings) ---
raw = [
    "Alice,30,Engineering",
    "Bob,25,Marketing",
    "Charlie,35,Engineering",
    "Diana,28,HR",
]

# Parse into list of dicts
employees = [
    {"name": parts[0], "age": int(parts[1]), "dept": parts[2]}
    for row in raw
    for parts in [row.split(",")]   # trick: bind intermediate value
]
print("Parsed employees:")
for emp in employees:
    print(" ", emp)

# Filter: only Engineering
eng_names = [e["name"] for e in employees if e["dept"] == "Engineering"]
print("\nEngineering team:", eng_names)

# Department → avg age (using dict comprehension + filter)
depts = {"Engineering", "Marketing", "HR"}
avg_age = {
    dept: sum(e["age"] for e in employees if e["dept"] == dept)
           // sum(1 for e in employees if e["dept"] == dept)
    for dept in depts
}
print("\nAvg age by dept:", avg_age)

# Unique departments as a set
unique_depts = {e["dept"] for e in employees}
print("Unique depts    :", unique_depts)
````

## Quick Reference

```python
# List — all squares
[x**2 for x in range(10)]

# List — filter only evens
[x for x in range(10) if x % 2 == 0]

# List — ternary (transform every item)
["even" if x % 2 == 0 else "odd" for x in range(5)]

# Nested — flatten 2D list
[num for row in matrix for num in row]

# Dict
{k: v for k, v in pairs}           # from pairs
{k: v for k, v in d.items() if v}  # filtered
{v: k for k, v in d.items()}       # inverted

# Set — unique values
{x.lower() for x in words}

# Generator — memory-efficient
sum(x**2 for x in range(1_000_000))
```

### When to use which
| Need | Use |
|------|-----|
| Build a list to reuse | List `[]` |
| Build a lookup table | Dict `{}` |
| Deduplicate | Set `{}` |
| One-time sum/max/min over large data | Generator `()` |
