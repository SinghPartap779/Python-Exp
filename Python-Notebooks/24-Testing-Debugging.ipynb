# Python Testing & Debugging Cheat Sheet

## 1. Debugging with print()

The simplest technique — add `print()` calls to inspect values at key points.

```python
def find_average(numbers):
    print(f"DEBUG: input = {numbers}")      # inspect input
    total = sum(numbers)
    print(f"DEBUG: total = {total}")        # inspect intermediate
    avg = total / len(numbers)
    print(f"DEBUG: avg = {avg}")            # inspect result
    return avg

find_average([10, 20, 30])
```

## 2. assert Statements

`assert condition, "message"` — raises `AssertionError` if the condition is False.

Great for catching bugs early during development.

```python
def divide(a, b):
    assert b != 0, "Denominator must not be zero"
    return a / b

print(divide(10, 2))   # 5.0
try:
    print(divide(10, 0))
except AssertionError as e:
    print(f"AssertionError: {e}")
```

## 3. unittest — Built-in Testing Framework

Write tests as methods inside a `TestCase` class. Each test method must start with `test_`.

```python
import unittest

# Code under test
def add(a, b): return a + b
def multiply(a, b): return a * b
def is_even(n): return n % 2 == 0

# Test class
class TestMathFunctions(unittest.TestCase):

    def test_add(self):
        self.assertEqual(add(2, 3), 5)
        self.assertEqual(add(-1, 1), 0)
        self.assertEqual(add(0, 0), 0)

    def test_multiply(self):
        self.assertEqual(multiply(3, 4), 12)
        self.assertEqual(multiply(0, 100), 0)
        self.assertEqual(multiply(-2, 5), -10)

    def test_is_even(self):
        self.assertTrue(is_even(4))
        self.assertFalse(is_even(7))
        self.assertTrue(is_even(0))

# Run tests in notebook
unittest.main(argv=[""], exit=False, verbosity=2)
```

## 4. Common unittest Assertions

| Method | Passes when |
|---|---|
| `assertEqual(a, b)` | `a == b` |
| `assertNotEqual(a, b)` | `a != b` |
| `assertTrue(x)` | `bool(x) is True` |
| `assertFalse(x)` | `bool(x) is False` |
| `assertIsNone(x)` | `x is None` |
| `assertIn(a, b)` | `a in b` |
| `assertRaises(Error)` | Block raises that error |
| `assertAlmostEqual(a, b)` | `a ≈ b` (for floats) |

```python
class TestExamples(unittest.TestCase):

    def test_string_upper(self):
        self.assertEqual("hello".upper(), "HELLO")

    def test_list_contains(self):
        self.assertIn(3, [1, 2, 3, 4])

    def test_raises_exception(self):
        with self.assertRaises(ZeroDivisionError):
            _ = 1 / 0

    def test_float_equality(self):
        self.assertAlmostEqual(0.1 + 0.2, 0.3, places=5)

unittest.main(argv=[""], exit=False, verbosity=2)
```

## 5. setUp and tearDown

Code that runs before/after each test method.

```python
class BankAccount:
    def __init__(self, balance=0):
        self.balance = balance
    def deposit(self, amount): self.balance += amount
    def withdraw(self, amount):
        if amount > self.balance: raise ValueError("Insufficient funds")
        self.balance -= amount

class TestBankAccount(unittest.TestCase):

    def setUp(self):             # runs before each test
        self.account = BankAccount(100)

    def tearDown(self): pass     # runs after each test (cleanup)

    def test_deposit(self):
        self.account.deposit(50)
        self.assertEqual(self.account.balance, 150)

    def test_withdraw(self):
        self.account.withdraw(30)
        self.assertEqual(self.account.balance, 70)

    def test_overdraft(self):
        with self.assertRaises(ValueError):
            self.account.withdraw(500)

unittest.main(argv=[""], exit=False, verbosity=2)
```

## 6. Debugging with pdb (Python Debugger)

Insert a breakpoint to step through code interactively.

```python
# Modern Python 3.7+ breakpoint
def buggy_sum(nums):
    total = 0
    for n in nums:
        # breakpoint()   # <-- uncomment to pause here
        total += n
    return total

print(buggy_sum([1, 2, 3]))  # Run without breakpoint for now
# pdb commands: n (next), s (step), c (continue), p var (print), q (quit)
```

### Challenge 1: Write Tests for a String Utilities Module

```python
# Functions to test
def reverse_string(s): return s[::-1]
def count_vowels(s): return sum(1 for c in s.lower() if c in "aeiou")
def is_palindrome(s): return s == s[::-1]

class TestStringUtils(unittest.TestCase):

    def test_reverse_string(self):
        self.assertEqual(reverse_string("hello"), "olleh")
        self.assertEqual(reverse_string(""), "")
        self.assertEqual(reverse_string("a"), "a")

    def test_count_vowels(self):
        self.assertEqual(count_vowels("hello"), 2)
        self.assertEqual(count_vowels("Python"), 1)
        self.assertEqual(count_vowels("rhythm"), 0)

    def test_is_palindrome(self):
        self.assertTrue(is_palindrome("racecar"))
        self.assertFalse(is_palindrome("hello"))
        self.assertTrue(is_palindrome("a"))

unittest.main(argv=[""], exit=False, verbosity=2)
```

### Challenge 2: Test-Driven Development (TDD) — Write tests first
Write failing tests first, then implement the function to make them pass.

```python
# Step 1: Write tests (they will fail until we implement the function)
class TestCalculator(unittest.TestCase):

    def test_factorial_base_cases(self):
        self.assertEqual(factorial(0), 1)
        self.assertEqual(factorial(1), 1)

    def test_factorial_normal(self):
        self.assertEqual(factorial(5), 120)
        self.assertEqual(factorial(10), 3628800)

    def test_factorial_negative(self):
        with self.assertRaises(ValueError):
            factorial(-1)

# Step 2: Implement the function
def factorial(n):
    if n < 0:
        raise ValueError("n must be non-negative")
    if n == 0:
        return 1
    return n * factorial(n - 1)

unittest.main(argv=[""], exit=False, verbosity=2)
```
