# Python Modules & Packages Cheat Sheet

## 1. What is a Module?

A **module** is any `.py` file. You can import it to reuse its code.

A **package** is a folder containing modules (and an `__init__.py`).

## 2. Import Styles

```python
import math                    # whole module
from math import sqrt, pi      # specific names
from math import sqrt as sq    # alias
import math as m               # module alias

print(math.factorial(5))
print(sqrt(16))
print(sq(9))
print(m.pi)
```

## 3. Creating Your Own Module

Save this as `greetings.py` (create it below), then import it.

```python
# Create greetings.py in the same directory
from pathlib import Path
Path("greetings.py").write_text("""
def hello(name):
    return f\"Hello, {name}!\"

def goodbye(name):
    return f\"Goodbye, {name}!\"

LANGUAGES = [\"English\", \"Spanish\", \"French\"]
"""
)
print("greetings.py created")
```

```python
import greetings
print(greetings.hello("Alice"))
print(greetings.goodbye("Bob"))
print(greetings.LANGUAGES)
```

## 4. __name__ == '__main__'

Code under this block runs only when the file is executed directly, not when imported.

```python
# Inside any module:
if __name__ == "__main__":
    # This only runs when you execute THIS file directly
    print("Running as a script")
else:
    print("Imported as a module")

# Current notebook context
print(__name__)  # __main__ in a notebook/script
```

## 6. Creating a Package

A package is a folder with an `__init__.py`.

```python
import os

# Create a mini package structure
os.makedirs("mypackage", exist_ok=True)

from pathlib import Path
Path("mypackage/__init__.py").write_text('from .utils import add, multiply')
Path("mypackage/utils.py").write_text("""
def add(a, b): return a + b
def multiply(a, b): return a * b
""")
print("Package created.")
```

```python
from mypackage import add, multiply
print(add(3, 4))        # 7
print(multiply(3, 4))   # 12
```

### Challenge 1: Build a mathtools module
Create `mathtools.py` with functions: `is_prime`, `factors`, `gcd`.

```python
from pathlib import Path
Path("mathtools.py").write_text("""
def is_prime(n):
    if n < 2: return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0: return False
    return True

def factors(n):
    return [i for i in range(1, n+1) if n % i == 0]

def gcd(a, b):
    while b:
        a, b = b, a % b
    return a
""")
print("mathtools.py created.")
```

```python
import mathtools
print(mathtools.is_prime(17))    # True
print(mathtools.factors(12))     # [1, 2, 3, 4, 6, 12]
print(mathtools.gcd(48, 18))     # 6
```

### Challenge 2: Explore random module
Use `random` to simulate rolling two dice 1000 times and show frequency of each total (2-12).

```python
import random
from collections import Counter

totals = [random.randint(1,6) + random.randint(1,6) for _ in range(1000)]
freq = Counter(totals)

print("Total | Count | Bar")
for total in range(2, 13):
    bar = "#" * (freq[total] // 5)
    print(f"  {total:2}  | {freq[total]:4}  | {bar}")
```

### Challenge 3: Use itertools
Use `itertools` to generate all combinations of a 4-item menu taken 2 at a time.

```python
from itertools import combinations, permutations

menu = ["pizza", "pasta", "salad", "soup"]

print("Combinations (order doesn't matter):")
for combo in combinations(menu, 2):
    print(" +".join(combo))

print("\nPermutations (order matters):")
for perm in permutations(menu, 2):
    print(" -> ".join(perm))
```
