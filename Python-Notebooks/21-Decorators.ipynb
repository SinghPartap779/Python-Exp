# Python Decorators Cheat Sheet

## 1. What is a Decorator?

A **decorator** is a function that wraps another function to add behaviour before/after it runs — without changing the original function.

Syntax: `@decorator_name` above a function definition.

## 2. Functions are First-Class Objects

Before decorators make sense, understand that functions can be passed around like variables.

```python
def greet():
    print("Hello!")

# Assign to a variable
say_hi = greet
say_hi()   # Hello!

# Pass as an argument
def run(func):
    func()

run(greet)  # Hello!

# Return from a function
def make_greeter():
    def inner():
        print("Hi from inner!")
    return inner

f = make_greeter()
f()   # Hi from inner!
```

## 3. Building a Decorator Manually

```python
def my_decorator(func):
    def wrapper():
        print("--- Before ---")
        func()
        print("--- After ---")
    return wrapper

def say_hello():
    print("Hello, World!")

# Apply decorator manually
decorated = my_decorator(say_hello)
decorated()
```

## 4. The @ Syntax (Syntactic Sugar)

`@decorator` is shorthand for `func = decorator(func)`.

```python
@my_decorator
def say_hello():
    print("Hello, World!")

say_hello()
# --- Before ---
# Hello, World!
# --- After ---
```

## 5. Decorators with Arguments — Using functools.wraps

```python
from functools import wraps

def my_decorator(func):
    @wraps(func)   # preserves func's name and docstring
    def wrapper(*args, **kwargs):
        print(f"Calling {func.__name__}")
        result = func(*args, **kwargs)
        print(f"Done")
        return result
    return wrapper

@my_decorator
def add(a, b):
    """Returns the sum of a and b."""
    return a + b

print(add(3, 5))
print(add.__name__)  # add (not 'wrapper' — thanks to @wraps)
```

## 6. Practical Decorator: Timer

```python
import time
from functools import wraps

def timer(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        start = time.perf_counter()
        result = func(*args, **kwargs)
        end = time.perf_counter()
        print(f"{func.__name__} took {end-start:.4f}s")
        return result
    return wrapper

@timer
def slow_sum(n):
    return sum(range(n))

print(slow_sum(1_000_000))
```

## 7. Practical Decorator: Retry on Failure

```python
import random
from functools import wraps

def retry(times=3):
    def decorator(func):
        @wraps(func)
        def wrapper(*args, **kwargs):
            for attempt in range(1, times + 1):
                try:
                    return func(*args, **kwargs)
                except Exception as e:
                    print(f"Attempt {attempt} failed: {e}")
            raise RuntimeError(f"Failed after {times} attempts")
        return wrapper
    return decorator

@retry(times=3)
def flaky_function():
    if random.random() < 0.7:   # 70% chance of failing
        raise ValueError("Random failure")
    return "Success!"

print(flaky_function())
```

## 8. Built-in Decorators

| Decorator | Purpose |
|---|---|
| `@staticmethod` | Method with no `self` or `cls` |
| `@classmethod` | Method that receives the class |
| `@property` | Getter — access method like an attribute |
| `@functools.lru_cache` | Cache return values (memoisation) |

```python
from functools import lru_cache

@lru_cache(maxsize=None)
def fib(n):
    if n < 2: return n
    return fib(n-1) + fib(n-2)

print([fib(i) for i in range(10)])  # very fast — results cached
print(fib.cache_info())
```

## Practice Challenges

### Challenge 1: Logger Decorator
Write a `@log_call` decorator that prints the function name, arguments, and return value.

```python
from functools import wraps

def log_call(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        print(f"Calling {func.__name__}({args}, {kwargs})")
        result = func(*args, **kwargs)
        print(f"  -> {result}")
        return result
    return wrapper

@log_call
def multiply(a, b):
    return a * b

@log_call
def greet(name, greeting="Hello"):
    return f"{greeting}, {name}!"

multiply(4, 5)
greet("Alice", greeting="Hi")
```

### Challenge 2: Validate Positive Arguments
Write a decorator that raises `ValueError` if any positional argument is negative.

```python
from functools import wraps

def require_positive(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        for arg in args:
            if isinstance(arg, (int, float)) and arg < 0:
                raise ValueError(f"All arguments must be non-negative, got {arg}")
        return func(*args, **kwargs)
    return wrapper

@require_positive
def area(width, height):
    return width * height

print(area(4, 5))   # 20
try:
    print(area(4, -5))
except ValueError as e:
    print(e)
```
