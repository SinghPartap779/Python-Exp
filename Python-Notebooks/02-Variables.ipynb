# Python Variables & Storage Cheat Sheet

## 1. Variable Assignment
Variables are created on assignment — no declaration needed. Python infers the type.

````python
name   = "Alice"     # str
age    = 25          # int
height = 5.6         # float
is_student = True    # bool

print(name, age, height, is_student)
print(type(name), type(age), type(height), type(is_student))
````

**What's happening here?**

- Each line creates a variable by simply writing `variable_name = value`. Python figures out the type automatically — no need to say `int age = 25` like in Java or C.
- `print(name, age, height, is_student)` prints all four values separated by spaces: `Alice 25 5.6 True`
- `type(x)` returns the **class** that `x` belongs to. For example, `type(25)` returns `<class 'int'>`.

> **Key idea:** In Python, a variable is just a *label* (name) that points to an object stored in memory. The object itself knows its own type.

## 2. Multiple & Chained Assignment
Assign to many variables in one line.

**What's happening here?**

- `x, y, z = 1, 2, 3` — This is called **tuple unpacking**. Python matches values left-to-right: `x=1`, `y=2`, `z=3`. The right side is a tuple `(1, 2, 3)` that gets "unpacked" into the variables on the left.
- `a = b = c = 0` — **Chained assignment** sets all three variables to the same value `0` in a single line.
- `a, b = b, a` — The **swap trick**: Python evaluates the entire right side `(b, a)` first (creating a temporary tuple `(20, 10)`), then assigns them to `a` and `b`. No temporary variable needed!

> **Why does the swap work?** Python evaluates the right-hand side completely before doing any assignment, so the old values of `a` and `b` are safely captured before anything is overwritten.

````python
# Assign different values at once
x, y, z = 1, 7, 3
print(x, y, z)   # 1 7 3

# Assign the same value to multiple variables
a = b = c = 5
print(a, b, c)   # 0 0 0

# Swap without a temp variable
a, b = 10, 20
a, b = b, a
print(a, b)      # 20 10
````

**What's happening here?**

- Each variable holds a different Python type. Here's a quick summary:
  | Variable | Value | Type | Notes |
  |---|---|---|---|
  | `i` | `42` | `int` | Whole number |
  | `f` | `3.14` | `float` | Decimal number |
  | `c` | `2+3j` | `complex` | `j` is the imaginary unit |
  | `s` | `"hello"` | `str` | Text, in quotes |
  | `b` | `True` | `bool` | Only `True` or `False` |
  | `n` | `None` | `NoneType` | Represents "nothing" / absence of value |
  | `lst` | `[1,2,3]` | `list` | Ordered, mutable (changeable) |
  | `tup` | `(1,2,3)` | `tuple` | Ordered, immutable (can't change after creation) |
  | `dct` | `{"k":"v"}` | `dict` | Key-value pairs |
  | `st` | `{1,2,3}` | `set` | Unique values, no duplicates |

- The loop line `f"{str(var):<20} -> {type(var).__name__}"` formats each output in a table-like style:
  - `:<20` pads the value string to 20 characters wide (left-aligned), so all `->` arrows line up.
  - `type(var).__name__` extracts just the type name as a plain string (e.g. `"int"` instead of `<class 'int'>`).

## 3. Data Types Overview
Python has several built-in types — all are objects.

**What's happening here?**

- `type(x) == str` — checks if the type is *exactly* `str`. Works, but has a pitfall with subclasses (explained below).
- `isinstance(x, str)` — **preferred way** to check types. Works correctly even if `x` is an instance of a class that *inherits* from `str`.
- **Type conversion (casting):** you can convert between types using built-in functions:
  - `int("42")` → converts the string `"42"` to the integer `42`
  - `float("3.14")` → converts string to decimal
  - `str(100)` → converts integer to its string representation `"100"`
  - `bool(0)` → `0` is **falsy**, so it becomes `False`
  - `bool("hello")` → any non-empty string is **truthy**, so it becomes `True`

> **Truthy vs Falsy:** In Python, some values are treated as `False` in a boolean context — these are `0`, `0.0`, `""` (empty string), `[]` (empty list), `{}` (empty dict), `None`. Everything else is treated as `True`.

> **Why prefer `isinstance` over `type(x) == str`?** If someone creates a custom class `class MyStr(str): ...`, then `type(obj) == str` would be `False` even though `obj` behaves like a string. `isinstance` correctly returns `True` in that case.

````python
#You can store the variables in a dictionary with names:
variables = {
    "i": 42,
    "f": 3.14,
    "c": 2 + 3j,
    "s": "hello",
    "b": True,
    "n": None,
    "lst": [1, 2, 3],
    "tup": (1, 2, 3),
    "dct": {"k": "v"},
    "st": {1, 2, 3}
}

for name, value in variables.items():
    print(f"{name:<5} {str(value):<20} -> {type(value).__name__}")
````

**What's happening here?**

- The variable `var` starts as an `int` (`10`), then gets reassigned to a `str` (`"hello"`), then to a `list` (`[1, 2, 3]`).
- In Python, this is perfectly valid! The variable `var` doesn't have a fixed type — it's just a name (label) that can point to any object.

> **Mental model — think of variables as sticky notes:**
> - The integer `10`, the string `"hello"`, and the list `[1, 2, 3]` are **objects** sitting somewhere in memory.
> - `var` is a sticky note that you peel off one object and stick onto another.
> - The old object may get deleted from memory if nothing else is pointing to it (Python does this automatically — it's called **garbage collection**).

> **Contrast with statically-typed languages** like Java or C++, where once you declare `int var = 10;`, `var` can only ever hold an integer. Python's dynamic typing gives more flexibility but requires you to be careful about what type a variable holds at any point in your program.

## 4. Type Checking & Conversion
`type()`, `isinstance()`, and casting functions.

**What's happening here?**

- **Valid names** — all of `my_var`, `_private`, `camelCase`, `var99` are legal Python identifiers.
- **Naming conventions** (not enforced by Python, but followed by everyone):
  - `snake_case` — words separated by underscores — is the standard for variables and functions.
  - `ALL_CAPS` — signals a constant (a value you don't intend to change). Python won't stop you from changing it, but it's a strong signal to other developers not to.
  - `PascalCase` (each word capitalized) — used for class names.
  - `_leading_underscore` — signals a "private" variable (an internal detail, not meant for outside use).
- `keyword.kwlist` prints all **reserved keywords** — words Python already uses for its syntax, like `if`, `for`, `while`, `return`, `class`, etc. You cannot use these as variable names because Python would get confused about your intent.

> **Common mistake:** `list = [1, 2, 3]` shadows Python's built-in `list` function, breaking it for the rest of the script. Avoid naming variables after built-ins like `list`, `dict`, `input`, `print`, `id`, `type`, `str`, `int`, `float`.

````python
x = "123"

# Check type
print(type(x))            # <class "str">
print(type(x) == str)     # True
print(isinstance(x, str)) # True  (preferred — works with inheritance)

# Type conversion (casting)
print(int("42"))          # 42
print(float("3.14"))      # 3.14
print(str(100))           # "100"
print(bool(0))            # False
print(bool("hello"))      # True  (non-empty string is truthy)
print(list((1, 2, 3)))    # [1, 2, 3]
print(tuple([4, 5, 6]))   # (4, 5, 6)
````

## 5. Dynamic Typing
A variable can point to a different type at any time.

**What's happening here?**

- `a = [1, 2, 3]` and `b = [1, 2, 3]` — two *separate* list objects in memory, both containing the same values.
- `c = a` — `c` does **not** create a copy of the list. It makes `c` point to the *exact same* list object as `a`. So `a`, and `c` are two names for one object.

| Expression | Result | Reason |
|---|---|---|
| `a == b` | `True` | Same values |
| `a is b` | `False` | Different objects in memory |
| `a is c` | `True` | Same object (c was assigned from a) |

- `id(x)` returns the **memory address** of the object `x`. Since `a` and `c` point to the same object, `id(a) == id(c)`. Since `b` is a different object, `id(b)` is different.

> **Small integer caching:** Python pre-creates integer objects for values `-5` to `256` and reuses them. So `x = 5; y = 5; x is y` is `True` because both names point to the *same cached object*. This is a Python implementation detail — don't rely on `is` for comparing values; always use `==` for value comparisons.

> **Practical impact of `c = a`:** If you do `c.append(4)`, the list `a` also changes! Because they are the same object. To make a true independent copy, use `c = a.copy()` or `c = list(a)`.

````python
var = 10
print(var, type(var))   # 10 <class "int">

var = "hello"
print(var, type(var))   # hello <class "str">

var = [1, 2, 3]
print(var, type(var))   # [1, 2, 3] <class "list">

# This is fine in Python — the variable is just a label
# pointing to an object in memory.
````

**What's happening here?**

The **LEGB rule** describes the order Python searches for a variable name:

```
Local → Enclosing → Global → Built-in
```

- **Local** — inside the current function. `inner()` has its own `x = "local"`.
- **Enclosing** — the function that contains the current function. `outer()` has `x = "enclosing"`, which `inner()` would see if it didn't define its own.
- **Global** — the module (file) level. `x = "global"` is here.
- **Built-in** — Python's own names like `print`, `len`, `range`, etc.

Step by step in this code:
1. `outer()` is called → `x = "enclosing"` is created in outer's local scope.
2. `inner()` is called from inside `outer()` → `x = "local"` is created in inner's own scope.
3. `inner` prints `"local"` (finds `x` in its own Local scope — stops searching there).
4. Back in `outer`, it prints `"enclosing"` (inner's `x` is gone, outer sees its own).
5. Back at the global level, it prints `"global"`.

> **Key point:** Each function has its own private scope. Creating `x = "local"` inside `inner()` does NOT change the `x` in `outer()` or the global `x`. They are three completely independent variables that just happen to share the same name.

**What's happening here?**

**`global` keyword:**
- By default, assigning inside a function creates a *new local* variable, even if a global with the same name exists.
- `global count` tells Python: "when I say `count` in this function, I mean the *global* `count`, not a new local one."
- Without `global count`, writing `count += 1` inside `increment()` would raise an `UnboundLocalError`.

**`nonlocal` keyword and closures:**
- `make_counter()` returns the inner function `step` — this is called a **closure**. `step` "closes over" the variable `n` from `make_counter`'s scope, keeping it alive even after `make_counter` has finished.
- `counter = make_counter()` runs `make_counter` and stores the returned `step` function. At this point `n = 0` is frozen inside the closure.
- `counter()` calls `step()`. `nonlocal n` lets `step` modify the `n` from the enclosing `make_counter` scope. Without it, `n += 1` would try to create a local `n` and fail.
- Each call to `counter()` increments the same `n`, so you get `1`, then `2`.

> **Why is this useful?** Closures let you create functions with private, persistent state — like a counter that remembers how many times it's been called — without using a class.

## 8. Memory & Identity
`id()` returns the memory address. `is` checks identity (same object), `==` checks equality (same value).

**How this works:** `global total` inside `add()` makes the function reach out to the module-level `total` instead of creating a new local variable. Each `add(n)` call modifies the same shared `total`. The final value is `0 + 10 + 25 + 5 = 40`.

````python
a = [1, 2, 3]
b = [1, 2, 3]
c = a

print(a == b)    # True  — same values
print(a is b)    # False — different objects in memory
print(a is c)    # True  — c points to the same object as a

print(id(a), id(b), id(c))

# Small integers and interned strings are cached by Python
x = 5
y = 5
print(x is y)    # True  — Python reuses small int objects (-5 to 256)
````

**How this works:** `list1` and `list2` are created independently with `[...]`, so Python allocates two separate objects in memory — even though they hold identical values. `==` compares contents (same → `True`), while `is` compares memory addresses (different objects → `False`). The `id()` calls confirm this by printing two different addresses.

## 9. Variable Scope
LEGB rule: **L**ocal → **E**nclosing → **G**lobal → **B**uilt-in

**How this works:** The three variables use type hints (`: str`, `: float`, `: bool`) to document what they hold. `f"₹{product_price:,.2f}"` formats the price with commas as thousands separators and exactly 2 decimal places — so `49999.99` becomes `₹49,999.99`. `in_stock` is a `bool`, so it prints as `True`.

````python
x = "global"

def outer():
    x = "enclosing"
    def inner():
        x = "local"
        print("inner :", x)   # local
    inner()
    print("outer :", x)       # enclosing

outer()
print("global:", x)           # global
````

````python
count = 0

def increment():
    global count        # declare intent to modify global
    count += 1

increment()
increment()
print(count)  # 2

def make_counter():
    n = 0
    def step():
        nonlocal n      # modify the enclosing scope variable
        n += 1
        return n
    return step

counter = make_counter()
print(counter())  # 1
print(counter())  # 2
````
