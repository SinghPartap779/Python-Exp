# Python Functions Cheat Sheet

## 1. What is a Function?

A **function** is a reusable block of code that performs a specific task.

It helps avoid repetition and makes programs easier to read and maintain.

## 2. Defining a Function

```python
def greet():
    print("Hello, welcome to Python!")

# def    -> keyword to define a function
# greet  -> function name
# ()     -> parentheses (can hold parameters)
# :      -> starts the function block
# indented code -> function body
```

## 3. Calling a Function

```python
greet()   # Output: Hello, welcome to Python!
```

## 4. Functions with Parameters

```python
def greet(name):
    print("Hello,", name)

greet("Alice")   # Output: Hello, Alice
```

## 5. Functions with Return Values

```python
def add(a, b):
    return a + b

result = add(5, 3)
print(result)   # Output: 8
```

## 6. Default Parameters

```python
def greet(name="Guest"):
    print("Hello,", name)

greet()          # Output: Hello, Guest
greet("Bob")     # Output: Hello, Bob
```

## 7. Keyword Arguments

```python
def introduce(name, age):
    print(f"My name is {name} and I am {age} years old.")

introduce(age=25, name="Charlie")
# Output: My name is Charlie and I am 25 years old.
```

## 8. Variable-Length Arguments

- `*args` -> multiple positional arguments
- `**kwargs` -> multiple keyword arguments

```python
def show_numbers(*args):
    for num in args:
        print(num)

show_numbers(1, 2, 3, 4)   # Prints each number
```

```python
def show_info(**kwargs):
    for key, value in kwargs.items():
        print(key, ":", value)

show_info(name="Alice", age=30)
# Output:
# name : Alice
# age : 30
```

## Practice Challenges

### Challenge 1: Factorial
Write a function to calculate the factorial of a number.

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

print(factorial(5))   # Output: 120
print(factorial(0))   # Output: 1
```

### Challenge 2: Palindrome Check
Create a function that checks if a string is a palindrome.

```python
def is_palindrome(s):
    s = s.lower().replace(" ", "")
    return s == s[::-1]

print(is_palindrome("racecar"))   # Output: True
print(is_palindrome("hello"))     # Output: False
print(is_palindrome("A man a plan a canal Panama"))  # Output: True
```

### Challenge 3: Maximum of a List
Write a function that takes a list of numbers and returns the maximum.

```python
def find_max(numbers):
    if not numbers:
        return None
    max_val = numbers[0]
    for num in numbers[1:]:
        if num > max_val:
            max_val = num
    return max_val

print(find_max([3, 7, 1, 9, 4]))   # Output: 9
print(find_max([-5, -1, -3]))       # Output: -1
```

### Challenge 4: Sum of Any Number of Arguments
Build a function that accepts any number of arguments and returns their sum.

```python
def total(*args):
    return sum(args)

print(total(1, 2, 3))         # Output: 6
print(total(10, 20, 30, 40))  # Output: 100
print(total(5))               # Output: 5
```

### Challenge 5: Multilingual Greeting
Create a function with default parameters for greeting someone in different languages.

```python
def greet_multilingual(name, language="English"):
    greetings = {
        "English": "Hello",
        "Spanish": "Hola",
        "French": "Bonjour",
        "German": "Hallo",
        "Japanese": "Konnichiwa",
    }
    greeting = greetings.get(language, "Hello")
    print(f"{greeting}, {name}!")

greet_multilingual("Alice")                 # Output: Hello, Alice!
greet_multilingual("Carlos", "Spanish")     # Output: Hola, Carlos!
greet_multilingual("Marie", "French")       # Output: Bonjour, Marie!
greet_multilingual("Yuki", "Japanese")      # Output: Konnichiwa, Yuki!
```
