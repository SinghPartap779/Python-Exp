# Python List Operations

A **list** is an ordered, mutable sequence. It can hold any mix of types — integers, strings, other lists, etc.  
Python tracks each element's **value** and **type** independently, which is why `count(3)` and `count("3")` return different results.

| Method | What it does | Mutates list? |
|--------|-------------|--------------|
| `append(x)` | Add one item to the end | Yes |
| `extend(iterable)` | Add all items from another iterable | Yes |
| `insert(i, x)` | Insert item at position `i` | Yes |
| `remove(x)` | Remove first occurrence of value `x` | Yes |
| `pop(i)` | Remove & return item at index `i` (default: last) | Yes |
| `count(x)` | Count occurrences of value `x` (type-sensitive) | No |
| `index(x)` | Return index of first `x` (type-sensitive) | No |
| `sort()` | Sort in-place | Yes |

## 1. `append` vs `extend` vs `insert` — Adding Items

````python
team = ["Alice", "Bob"]

# append — adds ONE item (even if that item is a list, it lands as a single element)
team.append("Charlie")
print("After append('Charlie'):", team)           # ['Alice', 'Bob', 'Charlie']

new_members = ["Diana", "Eve"]
team.append(new_members)                          # ← entire list becomes ONE nested element
print("After append(list)     :", team)           # ['Alice', 'Bob', 'Charlie', ['Diana', 'Eve']]

# extend — unpacks the iterable and adds each item individually
team2 = ["Alice", "Bob"]
team2.extend(new_members)
print("After extend(list)     :", team2)          # ['Alice', 'Bob', 'Diana', 'Eve']

# Key difference visualised
print("\nappend adds 1 element  → len:", len(["Alice", "Bob", new_members]))     # 3
print("extend adds N elements → len:", len(["Alice", "Bob"] + new_members))      # 4

# insert(index, value) — precise placement
scores = [10, 30, 40]
scores.insert(1, 20)          # put 20 at index 1, shift the rest right
print("\nAfter insert(1, 20):", scores)           # [10, 20, 30, 40]

scores.insert(0, 5)           # prepend
print("After insert(0, 5) :", scores)            # [5, 10, 20, 30, 40]

scores.insert(100, 99)        # index beyond end → appended safely
print("After insert(100,99):", scores)           # [5, 10, 20, 30, 40, 99]
````

## 2. `remove` vs `pop` — Removing Items

- **`remove(x)`** — searches by **value**, removes the first match. Raises `ValueError` if not found.  
- **`pop(i)`** — removes by **index** and **returns** the removed item. Default index is `-1` (last).

````python
queue = ["Alice", "Bob", "Alice", "Charlie"]

# remove — removes FIRST occurrence only
queue.remove("Alice")
print("After remove('Alice'):", queue)    # ['Bob', 'Alice', 'Charlie'] — second Alice survives

# Trying to remove something not in the list raises ValueError
try:
    queue.remove("Diana")
except ValueError as e:
    print("ValueError:", e)               # list.remove(x): x not in list

# pop — removes by index and RETURNS the value (useful when you need what was removed)
letters = ["a", "b", "c", "d", "e"]

last     = letters.pop()      # no index → removes last (-1)
first    = letters.pop(0)     # removes first
middle   = letters.pop(1)     # removes new index 1 (was "c")

print("\nPopped last  :", last)            # e
print("Popped first :", first)            # a
print("Popped middle:", middle)           # c
print("Remaining    :", letters)          # ['b', 'd']

# Classic stack (LIFO) with append + pop
stack = []
stack.append("task-1")
stack.append("task-2")
stack.append("task-3")
print("\nStack pop order:", stack.pop(), stack.pop(), stack.pop())  # task-3 task-2 task-1
````

## 3. `count` and `index` — Type-Sensitive Lookup

Python uses **value equality AND type** when searching a list.  
`3` (int) and `"3"` (str) are **not** the same element, even though they look identical when printed.

````python
mixed = [1, 2, 3, "3", 3, True, 3.0, "hello", 2, 2]

# count — how many times does this EXACT value (and type) appear?
print("count(3)     :", mixed.count(3))       # 3  — matches int 3, AND 3.0, AND True (see note below)
print('count("3")   :', mixed.count("3"))     # 1  — only the string "3"
print('count(2)     :', mixed.count(2))       # 3  — int 2 appears twice, plus... (see note)
print('count("hello"):', mixed.count("hello"))# 1

# Why True counts as 1 and 3.0 counts as 3:
# Python's == operator:  True == 1 → True,  3.0 == 3 → True
# count() uses ==, so these are considered equal in value.
# type() confirms they are different types:
print("\ntype(True) :", type(True))           # <class 'bool'>
print("True == 1  :", True == 1)             # True  ← same value by ==
print("True is 1  :", True is 1)             # False ← different objects

print("\ntype(3.0)  :", type(3.0))            # <class 'float'>
print("3.0 == 3   :", 3.0 == 3)             # True  ← same numeric value
print("3.0 == '3' :", 3.0 == "3")           # False ← int vs str never equal

# index — returns position of FIRST match (also type-sensitive via ==)
data = [10, 20, "20", 30, 20]
print("\nindex(20)  :", data.index(20))       # 1  — first int 20
print('index("20"):', data.index("20"))      # 2  — first string "20"

# index raises ValueError if not found — guard with 'in' first
if "missing" in data:
    print(data.index("missing"))
else:
    print("'missing' not in list — safe check before index()")

# Using index with start/stop to find the SECOND occurrence
print("\nSecond 20 at index:", data.index(20, 2))   # search from pos 2 onward → 4
````

## 4. `sort` — Sorting In-Place

`list.sort()` mutates the original list (returns `None`).  
`sorted(list)` returns a **new** sorted list without changing the original.

Key parameters: `reverse=True` for descending, `key=func` to sort by a custom rule.

````python
numbers = [42, 7, 19, 3, 88, 1]

# sort() mutates in-place — return value is None!
result = numbers.sort()
print("numbers after sort():", numbers)   # [1, 3, 7, 19, 42, 88]
print("return value        :", result)    # None  ← common mistake: assigning this loses the list

# reverse=True for descending
numbers.sort(reverse=True)
print("Descending          :", numbers)   # [88, 42, 19, 7, 3, 1]

# sorted() — non-destructive, original untouched
original = [5, 2, 9, 1]
new_list = sorted(original)
print("\noriginal after sorted():", original)  # [5, 2, 9, 1] — unchanged
print("new sorted list        :", new_list)   # [1, 2, 5, 9]

# key= : sort by a custom function (applied to each element before comparing)
words = ["banana", "fig", "apple", "kiwi", "date"]
words.sort(key=len)                            # sort by string length
print("\nSorted by length     :", words)        # ['fig', 'kiwi', 'date', 'apple', 'banana']

words.sort(key=str.lower)                      # case-insensitive alphabetical
print("Case-insensitive sort:", words)

# Sort a list of dicts by a specific field
employees = [
    {"name": "Charlie", "salary": 70000},
    {"name": "Alice",   "salary": 95000},
    {"name": "Bob",     "salary": 80000},
]
employees.sort(key=lambda e: e["salary"], reverse=True)
print("\nEmployees by salary (high→low):")
for emp in employees:
    print(f"  {emp['name']:10} ${emp['salary']:,}")
````

## 5. Values vs Types — How Python Distinguishes Elements

Python equality (`==`) compares **value**, not type — but `is` and `type()` expose the difference.  
This matters most in `count()`, `index()`, `in`, and `remove()`.

```
Value  Type     Python sees
──────────────────────────────
3      int      3
3.0    float    3.0  but  3.0 == 3  → True
"3"    str      "3"  and  "3" == 3  → False
True   bool     True but  True == 1 → True  (bool is a subclass of int)
```

````python
elements = [1, True, 1.0, "1", 1, False, 0, 0.0]

print("=== type() reveals the real type ===")
for val in elements:
    print(f"  value={val!r:6}  type={type(val).__name__}")

print("\n=== == (equality) vs is (identity) ===")
a, b, c = 1, True, 1.0
print(f"1 == True  : {a == b}")   # True  — same numeric value
print(f"1 is True  : {a is b}")   # False — different objects in memory
print(f"1 == 1.0   : {a == c}")   # True  — int and float equal
print(f"type(1)==type(1.0): {type(a)==type(c)}")  # False — int ≠ float

print("\n=== Practical impact on list methods ===")
bag = [1, True, 1.0, "1", 0, False]

# 'in' uses == under the hood
print("1     in bag:", 1     in bag)   # True (matches int 1)
print('"1"   in bag:', "1"   in bag)   # True (matches string "1")
print("True  in bag:", True  in bag)   # True (True == 1)
print("False in bag:", False in bag)   # True (False == 0)

# remove() deletes the FIRST == match regardless of type
bag_copy = bag.copy()
bag_copy.remove(1)         # removes the first element that == 1
print("\nAfter remove(1):", bag_copy)   # True and 1.0 survive because remove stops at first match

# Strict type check: use type(x) is int before operating
numbers_only = [v for v in bag if type(v) is int]
print("Strict int filter:", numbers_only)   # [1, 0]  — True/False excluded
````


## 2. deque (double-ended queue)
Fast O(1) insert/delete from **both ends**. `list.insert(0, x)` is O(n) — deque is faster.
**Use:** task queues · browser history · sliding window · BFS algorithms

````python
from collections import deque

# Browser history simulation
history = deque(maxlen=5)   # keeps only last 5 pages

for page in ['google.com', 'youtube.com', 'github.com', 'stackoverflow.com', 'reddit.com', 'openai.com']:
    history.append(page)
    
print("History:", list(history))   # google.com dropped (maxlen=5)

history.appendleft('homepage')     # add to front
print("After back:", list(history))
````

## 3. Counter
A `dict` subclass that **counts hashable objects**. Missing keys return 0 (not KeyError).
**Use:** word frequency · log analysis · vote counting · top-N items

````python
from collections import Counter

# Word frequency in a sentence
sentence = "the quick brown fox jumps over the lazy dog the fox"
words = sentence.split()

count = Counter(words)
print("All counts:", count)
print("Top 3 words:", count.most_common(3))
print("'the' appears:", count['the'], "times")
print("Missing key :", count['cat'])  # returns 0, not KeyError
````

## 4. OrderedDict
A dict that **explicitly preserves insertion order**. In Python 3.7+ regular `dict` also keeps order.
Still useful for: `move_to_end()` · `popitem(last=False)` (FIFO pop) · signalling order matters

````python
from collections import OrderedDict

# LRU-style cache using OrderedDict
cache = OrderedDict()
cache['page1'] = 'Home'
cache['page2'] = 'About'
cache['page3'] = 'Contact'

# Move recently accessed to end
cache.move_to_end('page1')
print("After access page1:", list(cache.keys()))

# Remove least recently used (front)
cache.popitem(last=False)
print("After eviction:", list(cache.keys()))
````

## 5. defaultdict
A dict that **never raises KeyError** — missing keys get an automatic default value.
**Use:** grouping records · counting · building adjacency lists · avoiding `if key in d` checks

````python
from collections import defaultdict

# Group students by department
students = [
    ('Alice', 'IT'), ('Bob', 'HR'), ('Charlie', 'IT'),
    ('Diana', 'Finance'), ('Eve', 'HR')
]

dept = defaultdict(list)
for name, department in students:
    dept[department].append(name)

print(dict(dept))
# No need for: if dept not in d: d[dept] = []
````

## 6. PySpark `.collect()` — Bringing DataFrame Data into a Python List

PySpark's `.collect()` is a **DataFrame action** that fetches ALL rows from a distributed DataFrame and returns them as a plain **Python list of `Row` objects** — which is why it belongs here alongside list operations.

| Method | Returns | Use when |
|--------|---------|----------|
| `df.show(n)` | `None` (just prints) | Quick visual preview |
| `df.collect()` | `list[Row]` — all rows | You need data in Python |
| `df.take(n)` | `list[Row]` — first n rows | You need a few rows in Python |
| `df.first()` | `Row` — single row | You need only the first row |

> **Warning:** `.collect()` pulls ALL data to the driver. Never use it on millions of rows — use `.take(n)` or `.limit(n).collect()` instead.

````python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("CollectDemo").getOrCreate()

# Sample DataFrame
data = [("Alice", 30), ("Bob", 25), ("Charlie", 35)]
df = spark.createDataFrame(data, ["Name", "Age"])

# --- collect() ---
rows = df.collect()
print("Type returned by collect():", type(rows))       # <class 'list'>
print("All rows:", rows)                               # [Row(Name='Alice', Age=30), ...]

# Access elements like a regular Python list
print("\nFirst row         :", rows[0])                # Row(Name='Alice', Age=30)
print("Name via key      :", rows[0]["Name"])          # Alice
print("Name via index    :", rows[0][0])               # Alice

# Iterate — just like any Python list
print("\n--- Iterate with for loop ---")
for row in rows:
    print(f"  {row['Name']} is {row['Age']} years old")

# Convert Row objects to plain Python dicts (easier to work with)
people = [row.asDict() for row in rows]
print("\nAs list of dicts:", people)

# Now apply normal list operations on the result
names = [row["Name"] for row in rows]
print("Names list       :", names)
print("Sorted names     :", sorted(names))

# --- take(n) — safer for large datasets ---
top2 = df.take(2)
print("\ntake(2):", top2)

# --- first() ---
print("first() :", df.first())

spark.stop()
````
