# Python File I/O Cheat Sheet

## 1. Opening a File

Use `open(path, mode)`. Always close the file after — or use a `with` block (recommended).

| Mode | Meaning |
|---|---|
| `'r'` | Read (default) |
| `'w'` | Write (overwrites) |
| `'a'` | Append |
| `'x'` | Create new (fails if exists) |
| `'b'` | Binary mode (e.g. `'rb'`) |
| `'+'` | Read + Write (e.g. `'r+'`) |

## 2. The with Statement (Context Manager)

Automatically closes the file even if an error occurs.

```python
with open("example.txt", "w") as f:
    f.write("Hello, File I/O!\n")
    f.write("Second line.\n")

# File is closed automatically here
print("File written.")
```

## 3. Reading a File

```python
# read() — entire file as one string
with open("example.txt", "r") as f:
    content = f.read()
print(content)
```

```python
# readlines() — list of lines (includes \n)
with open("example.txt", "r") as f:
    lines = f.readlines()
print(lines)
```

```python
# Iterate line by line (memory efficient)
with open("example.txt", "r") as f:
    for line in f:
        print(line.strip())
```

## 4. Writing to a File

- `write("w")` overwrites the entire file.
- `open("a")` appends without deleting existing content.

```python
# Append to existing file
with open("example.txt", "a") as f:
    f.write("Appended line.\n")

# writelines() — write a list of strings
lines = ["Line 1\n", "Line 2\n", "Line 3\n"]
with open("example.txt", "w") as f:
    f.writelines(lines)
```

## 5. Checking if a File Exists

```python
import os

if os.path.exists("example.txt"):
    print("File exists")
else:
    print("File not found")

# Safer: use pathlib (modern Python)
from pathlib import Path

path = Path("example.txt")
print(path.exists())   # True / False
print(path.stat().st_size, "bytes")
```

## 6. Working with CSV Files

```python
import csv

# Writing CSV
students = [
    ["Name", "Age", "Grade"],
    ["Alice", 20, "A"],
    ["Bob",   22, "B"],
    ["Charlie", 21, "A"],
]

with open("students.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(students)

print("CSV written.")
```

```python
# Reading CSV
with open("students.csv", "r") as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)
```

```python
# DictReader — each row as a dictionary
with open("students.csv", "r") as f:
    reader = csv.DictReader(f)
    for row in reader:
        print(row["Name"], "got", row["Grade"])
```

## 7. Working with JSON Files

```python
import json

data = {
    "name": "Alice",
    "age": 30,
    "hobbies": ["coding", "reading"]
}

# Write JSON
with open("data.json", "w") as f:
    json.dump(data, f, indent=4)

# Read JSON
with open("data.json", "r") as f:
    loaded = json.load(f)

print(loaded["name"])    # Alice
print(loaded["hobbies"]) # ['coding', 'reading']
```

## 8. pathlib — Modern File Paths

```python
from pathlib import Path

p = Path("example.txt")
print(p.name)       # example.txt
print(p.stem)       # example
print(p.suffix)     # .txt

# Read and write without open()
p.write_text("Hello from pathlib!")
print(p.read_text())
```

### Challenge 1: Word Counter
Read a text file and count the number of lines, words, and characters.

```python
def word_counter(path):
    try:
        with open(path, "r") as f:
            content = f.read()
        lines = content.splitlines()
        words = content.split()
        print(f"Lines: {len(lines)}")
        print(f"Words: {len(words)}")
        print(f"Characters: {len(content)}")
    except FileNotFoundError:
        print(f"File not found: {path}")

# Create a test file first
Path("sample.txt").write_text("Hello world\nPython is great\nFile IO is easy")
word_counter("sample.txt")
```

### Challenge 2: To-Do List Saved to File
Allow adding and viewing to-do items, persisted to a text file.

```python
TODO_FILE = "todos.txt"

def add_todo(task):
    with open(TODO_FILE, "a") as f:
        f.write(task.strip() + "\n")
    print(f"Added: {task}")

def view_todos():
    try:
        with open(TODO_FILE, "r") as f:
            tasks = f.readlines()
        if tasks:
            for i, task in enumerate(tasks, 1):
                print(f"{i}. {task.strip()}")
        else:
            print("No tasks yet.")
    except FileNotFoundError:
        print("No tasks yet.")

add_todo("Buy groceries")
add_todo("Read a book")
add_todo("Write Python code")
view_todos()
```

### Challenge 3: JSON Config File
Save and load a user config (name, theme, font_size) to/from JSON.

```python
import json

CONFIG_FILE = "config.json"

def save_config(config):
    with open(CONFIG_FILE, "w") as f:
        json.dump(config, f, indent=2)
    print("Config saved.")

def load_config():
    try:
        with open(CONFIG_FILE, "r") as f:
            return json.load(f)
    except FileNotFoundError:
        return {"name": "Guest", "theme": "light", "font_size": 12}

default = load_config()
print("Loaded:", default)

default["theme"] = "dark"
default["font_size"] = 14
save_config(default)
print("Updated:", load_config())
```

### Challenge 4: CSV Grade Reporter
Read a CSV of student scores and print a report showing pass/fail and the class average.

```python
import csv

# Create sample CSV
rows = [["Name","Score"],["Alice","88"],["Bob","45"],["Charlie","72"],["Diana","55"],["Eve","95"]]
with open("grades.csv","w",newline="") as f:
    csv.writer(f).writerows(rows)

# Read and report
total, count = 0, 0
with open("grades.csv", "r") as f:
    for row in csv.DictReader(f):
        score = int(row["Score"])
        status = "PASS" if score >= 50 else "FAIL"
        print(f"{row['Name']:10} {score:3}  {status}")
        total += score
        count += 1

print(f"\nClass average: {total/count:.1f}")
```
