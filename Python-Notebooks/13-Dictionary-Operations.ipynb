# Python Dictionary Operations Cheat Sheet

## 1. Creating Dictionaries
Dictionaries store **key-value pairs**. Keys must be unique and immutable.

```python
person = {"name": "Alice", "age": 25, "city": "Delhi"}
car = dict(brand="Toyota", year=2022, color="red")
scores = dict([("Math", 90), ("Science", 85)])
empty = {}

print(person)      # {"name": "Alice", "age": 25, "city": "Delhi"}
print(car)         # {"brand": "Toyota", "year": 2022, "color": "red"}
print(len(person)) # 3
```

## 2. Accessing Values
`[]`, `get()` — `get()` is safer; it never raises a KeyError.

```python
person = {"name": "Alice", "age": 25, "city": "Delhi"}

print(person["name"])              # Alice
print(person.get("age"))           # 25
print(person.get("email"))         # None  (no KeyError)
print(person.get("email", "N/A"))  # N/A   (default value)

# person["phone"]  ->  KeyError!
```

## 3. Adding & Updating Entries
`[]` assignment, `update()`

```python
person = {"name": "Alice", "age": 25}

person["city"] = "Delhi"   # add new key
person["age"]  = 26        # update existing key
print(person)

person.update({"email": "alice@example.com", "age": 27})
print(person)
```

## 4. Removing Entries
`del`, `pop()`, `popitem()`, `clear()`

```python
person = {"name": "Alice", "age": 25, "city": "Delhi", "email": "alice@example.com"}

del person["email"]              # delete by key (KeyError if missing)
print(person)

age = person.pop("age")          # remove & return value
print(age)                       # 25
print(person.pop("phone", 0))    # 0  (safe default)

last = person.popitem()          # remove & return last inserted (key, value)
print(last)

person.clear()                   # empty the dict
print(person)                    # {}
```

## 5. Iterating Over a Dictionary
`keys()`, `values()`, `items()`

```python
scores = {"Math": 90, "Science": 85, "English": 92}

for subject in scores.keys():
    print(subject)
print()

for mark in scores.values():
    print(mark)
print()

for subject, mark in scores.items():
    print(f"{subject}: {mark}")
```

## 6. Membership & Lookup
`in` checks keys by default. Use `.values()` to search values.

```python
person = {"name": "Alice", "age": 25}

print("name" in person)           # True
print("email" not in person)      # True
print(25 in person.values())      # True
print("Alice" in person.values()) # True
```

## 7. Useful Methods
`setdefault()`, `copy()`, `fromkeys()`

```python
# setdefault — insert with default only if key is absent
inventory = {"apple": 5}
inventory.setdefault("banana", 0)
inventory.setdefault("apple", 99)  # apple already exists, unchanged
print(inventory)  # {"apple": 5, "banana": 0}

# copy — shallow copy
original = {"x": 1, "y": 2}
clone = original.copy()
clone["z"] = 3
print(original)   # {"x": 1, "y": 2}  unaffected

# fromkeys — create dict from a list of keys with a shared default
keys = ["Math", "Science", "English"]
report = dict.fromkeys(keys, 0)
print(report)     # {"Math": 0, "Science": 0, "English": 0}
```

## 8. Merging Dictionaries
`update()`, `|` operator (Python 3.9+), `**` unpacking

```python
a = {"x": 1, "y": 2}
b = {"y": 99, "z": 3}   # "y" will be overwritten by b

# Method 1: update() — modifies a in place
a_copy = a.copy()
a_copy.update(b)
print(a_copy)   # {"x": 1, "y": 99, "z": 3}

# Method 2: | operator (Python 3.9+) — creates new dict
merged = a | b
print(merged)   # {"x": 1, "y": 99, "z": 3}

# Method 3: ** unpacking
merged2 = {**a, **b}
print(merged2)  # {"x": 1, "y": 99, "z": 3}
```

## 9. Dictionary Comprehension
Build dictionaries concisely in a single expression.

```python
squares = {x: x**2 for x in range(1, 6)}
print(squares)  # {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}

# Filter — keep only passing scores
scores = {"Alice": 85, "Bob": 42, "Carol": 91, "Dave": 38}
passed = {name: score for name, score in scores.items() if score >= 50}
print(passed)   # {"Alice": 85, "Carol": 91}

# Invert — swap keys and values
original = {"a": 1, "b": 2, "c": 3}
inverted = {v: k for k, v in original.items()}
print(inverted) # {1: "a", 2: "b", 3: "c"}
```

## 10. Nested Dictionaries
Dictionaries can hold other dictionaries as values.

```python
students = {
    "Alice": {"age": 20, "grade": "A"},
    "Bob":   {"age": 22, "grade": "B"},
}

print(students["Alice"]["grade"])   # A

students["Carol"] = {"age": 21, "grade": "A+"}

for name, info in students.items():
    print(f"{name} -> Age: {info['age']}, Grade: {info['grade']}")
```

## Practice Challenges

### Challenge 1: Word Frequency Counter
Count how many times each word appears in "the cat sat on the mat the cat".

```python
sentence = "the cat sat on the mat the cat"
freq = {}
for word in sentence.split():
    freq[word] = freq.get(word, 0) + 1
print(freq)  # {"the": 3, "cat": 2, "sat": 1, "on": 1, "mat": 1}
```

### Challenge 2: Invert a Dictionary
Swap keys and values in {"a": 1, "b": 2, "c": 3}.

```python
original = {"a": 1, "b": 2, "c": 3}
inverted = {v: k for k, v in original.items()}
print(inverted)  # {1: "a", 2: "b", 3: "c"}
```

### Challenge 3: Max Score
Find the student with the highest score from {"Alice": 85, "Bob": 92, "Carol": 78}.

```python
scores = {"Alice": 85, "Bob": 92, "Carol": 78}
topper = max(scores, key=scores.get)
print(f"Topper: {topper} with {scores[topper]} marks")  # Bob with 92
```

### Challenge 4: Merge & Override
Merge {"x": 1, "y": 2} and {"y": 10, "z": 3} so the second dict's values win.

```python
a = {"x": 1, "y": 2}
b = {"y": 10, "z": 3}
merged = {**a, **b}
print(merged)  # {"x": 1, "y": 10, "z": 3}
```

### Challenge 5: Group by First Letter
Group ["apple", "avocado", "banana", "blueberry", "cherry"] by their first letter.

```python
words = ["apple", "avocado", "banana", "blueberry", "cherry"]
grouped = {}
for word in words:
    grouped.setdefault(word[0], []).append(word)
print(grouped)
# {"a": ["apple", "avocado"], "b": ["banana", "blueberry"], "c": ["cherry"]}
```
