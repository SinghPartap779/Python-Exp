# Python DateTime Cheat Sheet

## 1. The datetime Module

Python's `datetime` module has four main classes:

| Class | What it holds |
|---|---|
| `date` | Year, month, day |
| `time` | Hour, minute, second, microsecond |
| `datetime` | Date + time combined |
| `timedelta` | A duration (difference between two datetimes) |

## 2. Creating date and datetime Objects

```python
from datetime import date, time, datetime, timedelta

# Today's date
today = date.today()
print(today)                 # e.g. 2024-06-15
print(today.year)            # 2024
print(today.month)           # 6
print(today.day)             # 15

# Specific date
birthday = date(1990, 3, 25)
print(birthday)              # 1990-03-25
```

```python
# Current datetime
now = datetime.now()
print(now)               # 2024-06-15 14:30:00.123456
print(now.hour)          # 14
print(now.minute)        # 30

# Specific datetime
dt = datetime(2024, 12, 25, 9, 0, 0)
print(dt)                # 2024-12-25 09:00:00
```

## 3. Formatting Dates — strftime()

Convert a datetime to a **string** using format codes.

| Code | Meaning | Example |
|---|---|---|
| `%Y` | 4-digit year | 2024 |
| `%m` | Month (01-12) | 06 |
| `%d` | Day (01-31) | 15 |
| `%H` | Hour 24h (00-23) | 14 |
| `%I` | Hour 12h (01-12) | 02 |
| `%M` | Minutes | 30 |
| `%S` | Seconds | 00 |
| `%p` | AM/PM | PM |
| `%A` | Full weekday | Saturday |
| `%B` | Full month | June |

```python
now = datetime.now()

print(now.strftime("%Y-%m-%d"))              # 2024-06-15
print(now.strftime("%d/%m/%Y"))              # 15/06/2024
print(now.strftime("%A, %B %d %Y"))          # Saturday, June 15 2024
print(now.strftime("%I:%M %p"))              # 02:30 PM
```

## 4. Parsing Strings — strptime()

Convert a **string** into a datetime object.

```python
date_str = "25/12/2024"
dt = datetime.strptime(date_str, "%d/%m/%Y")
print(dt)            # 2024-12-25 00:00:00
print(type(dt))      # <class 'datetime.datetime'>

time_str = "2024-06-15 14:30:00"
dt2 = datetime.strptime(time_str, "%Y-%m-%d %H:%M:%S")
print(dt2.weekday())  # 5 (Saturday, 0=Monday)
```

## 5. timedelta — Date Arithmetic

```python
today = date.today()

# Add/subtract days
tomorrow = today + timedelta(days=1)
yesterday = today - timedelta(days=1)
next_week = today + timedelta(weeks=1)

print(yesterday, today, tomorrow)
print(next_week)

# Difference between two dates
birthday = date(1990, 3, 25)
age_days = (today - birthday).days
print(f"You've been alive ~{age_days} days")
```

## 6. Comparing Datetimes

```python
from datetime import datetime

dt1 = datetime(2024, 1, 1)
dt2 = datetime(2024, 6, 15)

print(dt1 < dt2)    # True
print(dt1 == dt2)   # False
print(dt2 - dt1)    # 166 days, 0:00:00
```

### Challenge 1: Age Calculator
Calculate someone's age in years given their birthday.

```python
from datetime import date

def calculate_age(birthdate):
    today = date.today()
    years = today.year - birthdate.year
    # Subtract 1 if birthday hasn't happened yet this year
    if (today.month, today.day) < (birthdate.month, birthdate.day):
        years -= 1
    return years

bday = date(1990, 12, 25)
print(f"Age: {calculate_age(bday)} years")
```

### Challenge 2: Days Until Event
Calculate how many days until a future event.

```python
from datetime import date

def days_until(event_date):
    today = date.today()
    delta = event_date - today
    if delta.days < 0:
        return f"That date was {abs(delta.days)} days ago."
    elif delta.days == 0:
        return "That's today!"
    return f"{delta.days} days away."

print(days_until(date(2024, 12, 25)))   # Days until Christmas
print(days_until(date(2020, 1, 1)))     # Past date
```

### Challenge 3: Meeting Scheduler
Given a start time and duration in minutes, print the end time.

```python
from datetime import datetime, timedelta

def schedule_meeting(start_str, duration_minutes):
    start = datetime.strptime(start_str, "%Y-%m-%d %H:%M")
    end = start + timedelta(minutes=duration_minutes)
    print(f"Meeting: {start.strftime('%A %d %B %Y')}")
    print(f"  Start: {start.strftime('%H:%M')}")
    print(f"  End:   {end.strftime('%H:%M')}")

schedule_meeting("2024-06-20 09:30", 90)
```

### Challenge 4: Reformat a Log Timestamp
Convert log timestamps from US format to ISO 8601.

```python
from datetime import datetime

logs = [
    "06/15/2024 02:30:00 PM — Server started",
    "06/15/2024 02:31:45 PM — First request",
    "06/15/2024 03:00:00 PM — Health check",
]

for line in logs:
    parts = line.split(" — ")
    old_ts = parts[0].strip()
    dt = datetime.strptime(old_ts, "%m/%d/%Y %I:%M:%S %p")
    new_ts = dt.strftime("%Y-%m-%dT%H:%M:%S")
    print(f"{new_ts} — {parts[1]}")
```
