# Python User Input Cheat Sheet

## 1. What is User Input?

User input lets your program **pause and wait** for the user to type something.

Python reads it using the built-in `input()` function.

Everything entered is returned as a **string** — even if the user types a number.

## 2. The `input()` Function

```python
name = input("Enter your name: ")
print("Hello,", name)
```

## 3. Input is Always a String

`input()` **always returns a string**, even if the user types `42`.

```python
value = input("Enter a number: ")
print(type(value))   # <class 'str'>
print(value + value) # Concatenates, does NOT add numbers!
```

## 4. Converting Input Types

Use type conversion to work with numbers.

| Function | Converts to |
|---|---|
| `int(input())` | Integer |
| `float(input())` | Decimal number |
| `str(input())` | String (default) |
| `bool(input())` | Boolean |

```python
age = int(input("Enter your age: "))
print("Next year you will be", age + 1)
```

```python
price = float(input("Enter the price: "))
tax = price * 0.10
print(f"Tax: {tax:.2f}")
```

## 5. Multiple Inputs on One Line

Use `.split()` to read several values at once.

```python
# User types: 10 20 30
numbers = input("Enter three numbers separated by spaces: ").split()
print(numbers)          # ['10', '20', '30']

# Convert all to integers
a, b, c = map(int, input("Enter three numbers: ").split())
print(a + b + c)
```

## 6. Input Validation with try/except

If the user types something unexpected, the program can crash. Use `try/except` to handle it gracefully.

```python
try:
    num = int(input("Enter an integer: "))
    print("You entered:", num)
except ValueError:
    print("That was not a valid integer!")
```

## 7. Looping Until Valid Input

Keep asking until the user provides acceptable input.

```python
while True:
    try:
        age = int(input("Enter your age: "))
        if age < 0 or age > 120:
            print("Please enter a realistic age.")
        else:
            break
    except ValueError:
        print("Please enter a whole number.")

print("Age accepted:", age)
```

## 8. Yes/No Input

A common pattern for asking the user to confirm or choose.

```python
answer = input("Do you want to continue? (yes/no): ").strip().lower()

if answer in ("yes", "y"):
    print("Continuing...")
elif answer in ("no", "n"):
    print("Stopping.")
else:
    print("Unrecognised answer.")
```

### Challenge 1: Simple Calculator
Ask the user for two numbers and an operator (`+`, `-`, `*`, `/`), then print the result.

```python
def calculator():
    a = float(input("Enter first number: "))
    op = input("Enter operator (+, -, *, /): ").strip()
    b = float(input("Enter second number: "))

    if op == "+":
        print(f"Result: {a + b}")
    elif op == "-":
        print(f"Result: {a - b}")
    elif op == "*":
        print(f"Result: {a * b}")
    elif op == "/":
        if b == 0:
            print("Error: cannot divide by zero.")
        else:
            print(f"Result: {a / b}")
    else:
        print("Unknown operator.")

calculator()
```

### Challenge 2: Guess the Number
Pick a secret number. Ask the user to guess it, and tell them if their guess is too high, too low, or correct.

```python
import random

def guess_the_number():
    secret = random.randint(1, 20)
    attempts = 0

    print("Guess the number between 1 and 20!")
    while True:
        try:
            guess = int(input("Your guess: "))
        except ValueError:
            print("Please enter a whole number.")
            continue

        attempts += 1
        if guess < secret:
            print("Too low!")
        elif guess > secret:
            print("Too high!")
        else:
            print(f"Correct! You got it in {attempts} attempt(s).")
            break

guess_the_number()
```

### Challenge 3: Sum of N Numbers
Ask the user how many numbers they want to enter, collect each one, then print the total and average.

```python
def sum_of_n():
    while True:
        try:
            n = int(input("How many numbers? "))
            if n <= 0:
                print("Please enter a positive number.")
            else:
                break
        except ValueError:
            print("Please enter a whole number.")

    numbers = []
    for i in range(1, n + 1):
        while True:
            try:
                num = float(input(f"  Number {i}: "))
                numbers.append(num)
                break
            except ValueError:
                print("Invalid — try again.")

    total = sum(numbers)
    average = total / n
    print(f"Total: {total}")
    print(f"Average: {average:.2f}")

sum_of_n()
```

### Challenge 4: Password Checker
Ask the user to create a password. Validate that it is at least 8 characters long and contains at least one digit.

```python
def is_strong_password(pwd):
    if len(pwd) < 8:
        return False, "Password must be at least 8 characters."
    if not any(ch.isdigit() for ch in pwd):
        return False, "Password must contain at least one digit."
    return True, "Password accepted!"

while True:
    password = input("Create a password: ")
    valid, message = is_strong_password(password)
    print(message)
    if valid:
        break
```

### Challenge 5: Shopping List Builder
Let the user add items to a shopping list by typing them one at a time. Stop when they type `done`, then print the full list.

```python
def build_shopping_list():
    shopping_list = []
    print("Enter items one at a time. Type 'done' to finish.")

    while True:
        item = input("Add item: ").strip()
        if item.lower() == "done":
            break
        if item:
            shopping_list.append(item)
            print(f"  '{item}' added.")
        else:
            print("  Empty input ignored.")

    if shopping_list:
        print("\nYour shopping list:")
        for i, item in enumerate(shopping_list, 1):
            print(f"  {i}. {item}")
    else:
        print("Your list is empty.")

build_shopping_list()
```
